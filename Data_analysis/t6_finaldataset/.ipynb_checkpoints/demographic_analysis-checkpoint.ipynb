# T6 — Demographic Analysis

Evaluates whether **shoe size**, **age**, or **gender** produced different cadence responses
to the tempo manipulation, overall and split by test configuration.

**Reads from existing exports** — run the main `pressure_timeseries.ipynb` first to generate:
- `exports/initial_tempo_analysis/initial_tempo_stats.csv`
- `exports/figure5/influence_table.csv`

**Outcome variables:**
- `pct_change` — % cadence change from baseline to trial end
- `influence_score` — trajectory-based coupling strength (0–1)
- `correct_direction` — whether cadence moved in the target direction
- `agency_aggregate` — perceived sense of agency (0–10)

**Configurations tested:** direction (speeding up / slowing down) × condition order (tempo first / weight first)

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import ipywidgets as widgets
from IPython.display import display
from scipy import stats
from pathlib import Path

%matplotlib inline
plt.rcParams.update({'figure.dpi': 120, 'font.size': 10})

DATA_DIR = Path('.')

# ── Colour scheme ─────────────────────────────────────────────────────────────
GENDER_COLORS  = {'Female': '#e74c3c', 'Male': '#3498db'}
DIR_COLORS     = {'speeding up': '#3498db', 'slowing down': '#e74c3c'}
ORDER_COLORS   = {'tempo first': '#27ae60', 'weight first': '#e67e22'}

print('Setup complete.')
```

```python
# ── Load exported results ─────────────────────────────────────────────────────
df_tempo = pd.read_csv(DATA_DIR / 'exports/initial_tempo_analysis/initial_tempo_stats.csv')
df_tempo['pid'] = df_tempo['pid'].astype(str).str.zfill(2)

df_inf = pd.read_csv(DATA_DIR / 'exports/figure5/influence_table.csv', encoding='latin-1')
df_inf['pid'] = df_inf['pid'].astype(str).str.zfill(2)

# ── Load demographics from summary spreadsheet ────────────────────────────────
df_demo = pd.read_excel(DATA_DIR / 'Test 6_summary.xlsx', header=2)
df_demo['ID'] = df_demo['ID'].astype(str).str.zfill(2)
df_demo = df_demo[['ID', 'Gender', 'Age', 'UK shoe']].dropna(subset=['ID'])
df_demo.columns = ['pid', 'gender', 'age', 'shoe_size']
df_demo['shoe_size'] = pd.to_numeric(df_demo['shoe_size'], errors='coerce')
df_demo['age']       = pd.to_numeric(df_demo['age'],       errors='coerce')

# ── Merge all ─────────────────────────────────────────────────────────────────
df = df_tempo.merge(df_inf[['pid','influence_score','profile','tier']], on='pid', how='left')
df = df.merge(df_demo, on='pid', how='left')
df = df.dropna(subset=['gender','age','shoe_size'])

print(f'Merged dataset: {len(df)} sessions')
print(f"Gender: {df['gender'].value_counts().to_dict()}")
print(f"Age:    {df['age'].min():.0f}–{df['age'].max():.0f}  (median {df['age'].median():.0f})")
print(f"Shoe:   {df['shoe_size'].min()}–{df['shoe_size'].max()}  (median {df['shoe_size'].median()})")
print(f"Direction split: {df['direction'].value_counts().to_dict()}")
display(df[['pid','gender','age','shoe_size','direction','order',
            'pct_change','influence_score','correct_direction']].sort_values('pid'))
```

## Gender

```python
OUTCOMES = [
    ('pct_change',      '% cadence change'),
    ('influence_score', 'Influence score'),
]
CONFIGS = [
    ('All',              df),
    ('Speeding up',      df[df['direction']=='speeding up']),
    ('Slowing down',     df[df['direction']=='slowing down']),
    ('Tempo first',      df[df['order']=='tempo first']),
    ('Weight first',     df[df['order']=='weight first']),
]

fig, axes = plt.subplots(2, len(CONFIGS), figsize=(3.5*len(CONFIGS), 8),
                          sharey=False)
fig.subplots_adjust(hspace=0.45, wspace=0.3)

for row, (outcome, ylabel) in enumerate(OUTCOMES):
    for col, (cfg_name, cfg_df) in enumerate(CONFIGS):
        ax = axes[row][col]
        F = cfg_df[cfg_df['gender']=='Female'][outcome].dropna()
        M = cfg_df[cfg_df['gender']=='Male'][outcome].dropna()

        bp = ax.boxplot([F.values, M.values], positions=[0,1], widths=0.45,
                        patch_artist=True,
                        medianprops=dict(color='black',lw=2),
                        whiskerprops=dict(lw=1), capprops=dict(lw=1),
                        flierprops=dict(marker='o',ms=3,alpha=0.5))
        for patch, color in zip(bp['boxes'], [GENDER_COLORS['Female'],GENDER_COLORS['Male']]):
            patch.set_facecolor(color); patch.set_alpha(0.45)

        rng = np.random.default_rng(42)
        for xi, (vals, color) in enumerate([(F,GENDER_COLORS['Female']),(M,GENDER_COLORS['Male'])]):
            jit = rng.uniform(-0.12,0.12,len(vals))
            ax.scatter(xi+jit, vals, color=color, s=18, alpha=0.7, zorder=3)

        # Mann-Whitney U
        if len(F)>1 and len(M)>1:
            _, p = stats.mannwhitneyu(F, M, alternative='two-sided')
            sig = '***' if p<0.001 else ('**' if p<0.01 else ('*' if p<0.05 else 'ns'))
            y_top = max(F.max() if len(F) else 0, M.max() if len(M) else 0)
            ax.plot([0,1],[y_top*1.05, y_top*1.05], color='black', lw=0.8)
            ax.text(0.5, y_top*1.08, f'{sig} p={p:.3f}',
                    ha='center', va='bottom', fontsize=7)

        ax.set_xticks([0,1])
        ax.set_xticklabels([f'F (n={len(F)})', f'M (n={len(M)})'], fontsize=8)
        if col==0: ax.set_ylabel(ylabel, fontsize=9)
        if row==0: ax.set_title(cfg_name, fontsize=9)
        ax.grid(axis='y', linestyle='--', alpha=0.3)
        ax.axhline(0, color='#bdc3c7', lw=0.7)

fig.suptitle('Gender × outcome across configurations', fontsize=12)
plt.show()

# Summary stats table
rows = []
for outcome, ylabel in OUTCOMES:
    for cfg_name, cfg_df in CONFIGS:
        F = cfg_df[cfg_df['gender']=='Female'][outcome].dropna()
        M = cfg_df[cfg_df['gender']=='Male'][outcome].dropna()
        if len(F)>1 and len(M)>1:
            _, p = stats.mannwhitneyu(F, M, alternative='two-sided')
        else:
            p = float('nan')
        rows.append({'outcome':ylabel,'config':cfg_name,
                     'F median':round(float(F.median()),3) if len(F)>0 else None,
                     'M median':round(float(M.median()),3) if len(M)>0 else None,
                     'p (MWU)':round(p,4), 'sig':'*' if p<0.05 else ''})
display(pd.DataFrame(rows).set_index(['outcome','config']))
```

## Age

```python
fig, axes = plt.subplots(2, len(CONFIGS), figsize=(3.5*len(CONFIGS), 7),
                          sharey=False)
fig.subplots_adjust(hspace=0.45, wspace=0.3)

for row, (outcome, ylabel) in enumerate(OUTCOMES):
    for col, (cfg_name, cfg_df) in enumerate(CONFIGS):
        ax = axes[row][col]
        sub = cfg_df[['age', outcome, 'direction']].dropna()

        for direction, color in DIR_COLORS.items():
            s = sub[sub['direction']==direction]
            if len(s)<2: continue
            ax.scatter(s['age'], s[outcome], color=color, s=30, alpha=0.7,
                       label=direction, edgecolors='white', linewidths=0.4)

        if len(sub)>2:
            rho, p = stats.spearmanr(sub['age'], sub[outcome])
            sig = '*' if p<0.05 else ''
            ax.text(0.03, 0.96, f'ρ={rho:+.2f}{sig}\np={p:.3f}',
                    transform=ax.transAxes, fontsize=7.5, va='top',
                    bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.8))

        ax.axhline(0, color='#bdc3c7', lw=0.7)
        ax.set_xlabel('Age', fontsize=8)
        if col==0: ax.set_ylabel(ylabel, fontsize=9)
        if row==0: ax.set_title(cfg_name, fontsize=9)
        if row==0 and col==0: ax.legend(fontsize=7, loc='upper right')
        ax.grid(linestyle='--', alpha=0.3)

fig.suptitle('Age × outcome (Spearman ρ)', fontsize=12)
plt.show()

# Summary
rows = []
for outcome, ylabel in OUTCOMES:
    for cfg_name, cfg_df in CONFIGS:
        sub = cfg_df[['age', outcome]].dropna()
        if len(sub)>2:
            rho, p = stats.spearmanr(sub['age'], sub[outcome])
        else:
            rho, p = float('nan'), float('nan')
        rows.append({'outcome':ylabel,'config':cfg_name,
                     'n':len(sub),'Spearman ρ':round(rho,3),'p':round(p,4),
                     'sig':'*' if p<0.05 else ''})
display(pd.DataFrame(rows).set_index(['outcome','config']))
```

## Shoe size

UK shoe size used as a proxy for foot size. May correlate with stride characteristics and natural cadence variability.

```python
fig, axes = plt.subplots(2, len(CONFIGS), figsize=(3.5*len(CONFIGS), 7),
                          sharey=False)
fig.subplots_adjust(hspace=0.45, wspace=0.3)

for row, (outcome, ylabel) in enumerate(OUTCOMES):
    for col, (cfg_name, cfg_df) in enumerate(CONFIGS):
        ax = axes[row][col]
        sub = cfg_df[['shoe_size', outcome, 'direction', 'gender']].dropna()

        for direction, color in DIR_COLORS.items():
            s = sub[sub['direction']==direction]
            if len(s)<2: continue
            ax.scatter(s['shoe_size'], s[outcome], color=color, s=30, alpha=0.7,
                       label=direction, edgecolors='white', linewidths=0.4)

        if len(sub)>2:
            rho, p = stats.spearmanr(sub['shoe_size'], sub[outcome])
            sig = '*' if p<0.05 else ''
            ax.text(0.03, 0.96, f'ρ={rho:+.2f}{sig}\np={p:.3f}',
                    transform=ax.transAxes, fontsize=7.5, va='top',
                    bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.8))

        ax.axhline(0, color='#bdc3c7', lw=0.7)
        ax.set_xlabel('UK shoe size', fontsize=8)
        if col==0: ax.set_ylabel(ylabel, fontsize=9)
        if row==0: ax.set_title(cfg_name, fontsize=9)
        if row==0 and col==0: ax.legend(fontsize=7, loc='upper right')
        ax.grid(linestyle='--', alpha=0.3)

fig.suptitle('Shoe size × outcome (Spearman ρ)', fontsize=12)
plt.show()

# Summary
rows = []
for outcome, ylabel in OUTCOMES:
    for cfg_name, cfg_df in CONFIGS:
        sub = cfg_df[['shoe_size', outcome]].dropna()
        if len(sub)>2:
            rho, p = stats.spearmanr(sub['shoe_size'], sub[outcome])
        else:
            rho, p = float('nan'), float('nan')
        rows.append({'outcome':ylabel,'config':cfg_name,
                     'n':len(sub),'Spearman ρ':round(rho,3),'p':round(p,4),
                     'sig':'*' if p<0.05 else ''})
display(pd.DataFrame(rows).set_index(['outcome','config']))
```

## Summary overview

Correlation matrix of all demographic variables against all outcome variables, across all sessions.

```python
df['gender_num'] = (df['gender']=='Male').astype(int)  # 0=Female, 1=Male

DEMO_VARS    = [('age','Age'), ('shoe_size','Shoe size'), ('gender_num','Gender (M=1)')]
OUTCOME_VARS = [('pct_change','% cadence change'), ('influence_score','Influence score'),
                ('correct_direction','Correct direction (0/1)')]
df['correct_direction'] = df['correct_direction'].astype(float)

# ── Heatmap of Spearman correlations ─────────────────────────────────────────
rho_mat = np.zeros((len(DEMO_VARS), len(OUTCOME_VARS)))
p_mat   = np.ones_like(rho_mat)

for i, (dv, _) in enumerate(DEMO_VARS):
    for j, (ov, _) in enumerate(OUTCOME_VARS):
        sub = df[[dv, ov]].dropna()
        if len(sub)>2:
            rho_mat[i,j], p_mat[i,j] = stats.spearmanr(sub[dv], sub[ov])

fig, ax = plt.subplots(figsize=(8, 3.5))
im = ax.imshow(rho_mat, cmap='RdBu_r', vmin=-0.6, vmax=0.6, aspect='auto')
plt.colorbar(im, ax=ax, label='Spearman ρ')
ax.set_xticks(range(len(OUTCOME_VARS)))
ax.set_yticks(range(len(DEMO_VARS)))
ax.set_xticklabels([o for _,o in OUTCOME_VARS], fontsize=9)
ax.set_yticklabels([d for _,d in DEMO_VARS], fontsize=9)

for i in range(len(DEMO_VARS)):
    for j in range(len(OUTCOME_VARS)):
        sig = '*' if p_mat[i,j] < 0.05 else ''
        ax.text(j, i, f'{rho_mat[i,j]:+.2f}{sig}',
                ha='center', va='center', fontsize=9,
                color='white' if abs(rho_mat[i,j])>0.35 else 'black')

ax.set_title('Spearman ρ: demographics × outcomes  (* p<0.05)', fontsize=11)
plt.tight_layout()
plt.show()

# ── Full significance table ────────────────────────────────────────────────────
print('\n=== Full Spearman correlation table ===')
rows = []
for dv, dlabel in DEMO_VARS:
    for ov, olabel in OUTCOME_VARS:
        for cfg_name, cfg_df in CONFIGS:
            sub = cfg_df[[dv, ov]].dropna()
            if len(sub)>2:
                rho, p = stats.spearmanr(sub[dv], sub[ov])
            else:
                rho, p = float('nan'), float('nan')
            rows.append({'demographic':dlabel,'outcome':olabel,'config':cfg_name,
                         'n':len(sub),'ρ':round(rho,3),'p':round(p,4),
                         'sig':'* ' if p<0.05 else ''})
df_results = pd.DataFrame(rows)
sig_only = df_results[df_results['sig']=='* ']
if len(sig_only):
    print('\nSignificant results (p<0.05):')
    display(sig_only.set_index(['demographic','outcome','config']))
else:
    print('No significant results found (p<0.05).')
print('\nFull table:')
display(df_results.set_index(['demographic','outcome','config']))
```

## Export

```python
from pathlib import Path

export_btn    = widgets.Button(description='Export all', button_style='info',
                                layout=widgets.Layout(width='130px'))
export_status = widgets.Label('')

def do_export(_):
    d = Path('exports/demographic_analysis')
    d.mkdir(parents=True, exist_ok=True)
    export_status.value = 'Exporting…'

    # Merged dataset
    df.drop(columns=['gender_num'], errors='ignore').to_csv(d/'merged_demo_results.csv', index=False)

    # Full correlation table
    df_results.to_csv(d/'spearman_correlations.csv', index=False)

    # Gender figure
    fig_g, axes_g = plt.subplots(2, len(CONFIGS), figsize=(3.5*len(CONFIGS), 8),
                                  sharey=False)
    fig_g.subplots_adjust(hspace=0.45, wspace=0.3)
    rng = np.random.default_rng(42)
    for row, (outcome, ylabel) in enumerate(OUTCOMES):
        for col, (cfg_name, cfg_df) in enumerate(CONFIGS):
            ax = axes_g[row][col]
            F = cfg_df[cfg_df['gender']=='Female'][outcome].dropna()
            M = cfg_df[cfg_df['gender']=='Male'][outcome].dropna()
            if len(F)==0 and len(M)==0: continue
            bp = ax.boxplot([F.values if len(F) else [0], M.values if len(M) else [0]],
                            positions=[0,1], widths=0.45, patch_artist=True,
                            medianprops=dict(color='black',lw=2),
                            whiskerprops=dict(lw=1), capprops=dict(lw=1),
                            flierprops=dict(marker='o',ms=3,alpha=0.5))
            for patch, color in zip(bp['boxes'],[GENDER_COLORS['Female'],GENDER_COLORS['Male']]):
                patch.set_facecolor(color); patch.set_alpha(0.45)
            for xi, (vals, color) in enumerate([(F,GENDER_COLORS['Female']),(M,GENDER_COLORS['Male'])]):
                if len(vals)==0: continue
                jit = rng.uniform(-0.12,0.12,len(vals))
                ax.scatter(xi+jit, vals, color=color, s=18, alpha=0.7, zorder=3)
            if len(F)>1 and len(M)>1:
                _, p = stats.mannwhitneyu(F, M, alternative='two-sided')
                sig = '***' if p<0.001 else ('**' if p<0.01 else ('*' if p<0.05 else 'ns'))
                y_top = max(F.max(), M.max())
                ax.plot([0,1],[y_top*1.05,y_top*1.05], color='black', lw=0.8)
                ax.text(0.5, y_top*1.08, f'{sig} p={p:.3f}', ha='center', va='bottom', fontsize=7)
            ax.set_xticks([0,1])
            ax.set_xticklabels([f'F (n={len(F)})',f'M (n={len(M)})'], fontsize=8)
            if col==0: ax.set_ylabel(ylabel, fontsize=9)
            if row==0: ax.set_title(cfg_name, fontsize=9)
            ax.grid(axis='y', linestyle='--', alpha=0.3)
    fig_g.suptitle('Gender × outcome across configurations', fontsize=12)
    plt.tight_layout()
    fig_g.savefig(d/'gender_analysis.png', dpi=150, bbox_inches='tight')
    plt.close(fig_g)

    # Heatmap
    fig_h, ax_h = plt.subplots(figsize=(8, 3.5))
    im = ax_h.imshow(rho_mat, cmap='RdBu_r', vmin=-0.6, vmax=0.6, aspect='auto')
    plt.colorbar(im, ax=ax_h, label='Spearman ρ')
    ax_h.set_xticks(range(len(OUTCOME_VARS)))
    ax_h.set_yticks(range(len(DEMO_VARS)))
    ax_h.set_xticklabels([o for _,o in OUTCOME_VARS], fontsize=9)
    ax_h.set_yticklabels([d2 for _,d2 in DEMO_VARS], fontsize=9)
    for i in range(len(DEMO_VARS)):
        for j in range(len(OUTCOME_VARS)):
            sig = '*' if p_mat[i,j]<0.05 else ''
            ax_h.text(j,i,f'{rho_mat[i,j]:+.2f}{sig}',ha='center',va='center',fontsize=9,
                      color='white' if abs(rho_mat[i,j])>0.35 else 'black')
    ax_h.set_title('Spearman ρ: demographics × outcomes  (* p<0.05)', fontsize=11)
    plt.tight_layout()
    fig_h.savefig(d/'correlation_heatmap.png', dpi=150, bbox_inches='tight')
    plt.close(fig_h)

    export_status.value = f'✓ Saved to exports/demographic_analysis/'
    print(f'Saved: merged_demo_results.csv, spearman_correlations.csv, gender_analysis.png, correlation_heatmap.png')

export_btn.on_click(do_export)
display(widgets.HBox([export_btn]), export_status)
```
