# T6 — Weight Gait Analysis
## Replication of Tajadura-Jiménez et al. (2015)

**Reference:** Tajadura-Jiménez, A., Basia, M., Deroy, O., Fairhurst, M., Marquardt, N., & Bianchi-Berthouze, N. (2015). *As Light as your Footsteps: Altering Walking Sounds to Change Perceived Body Weight, Emotional State and Gait.* CHI 2015.

---

### What the paper found

The paper manipulated the frequency spectrum of participants' self-produced walking sounds in real-time:
- **Low frequency** condition (heavier sound): more energy below 250 Hz → perceived heavier body
- **High frequency** condition (lighter sound): more energy above 1 kHz → perceived lighter body

**Primary gait findings (Figure 4):**
1. **Heel contact time** (FSR): *longer* in the Low frequency (heavier) condition — t(20)=2.79, p=0.006
2. **Upward foot acceleration** (accelerometer): *larger* in the High frequency (lighter) condition — t(20)=2.62, p=0.008

The paper z-scored all physiological data within participants to reduce intra-subject variability, then compared conditions with Shapiro-Wilk normality testing followed by ANOVA / Wilcoxon tests.

---

### Test 6 analog

| Paper | Test 6 |
|---|---|
| Low frequency (heavier sound) | `increasing` direction, wc plateau (+12) |
| High frequency (lighter sound) | `decreasing` direction, wc plateau (−12) |
| Control (natural sound) | Baseline segment (|wc| ≤ 0.5) within same session |
| FSR heel contact time | `detect_contacts()` on pressure insole L/R heel |
| Upward foot acceleration | Peak net acceleration √(ax²+ay²+az²) during swing phase (synced IMU) |
| Per-trial step average | Per-segment mean across all steps at baseline or plateau wc |
| Within-participant z-scores | Z-score all steps within participant's weight session |

**Excluded:** P01, P06 (BLE timestamp fragmentation — unusable IMU data)

> **Note on data transform:** Cadence (SPM) in the tempo analysis is already a normalised rate measure with low inter-participant variance, so raw values are compared directly. FSR heel-contact and IMU acceleration data carry large between-person offsets, so within-participant z-scores are applied here following Tajadura-Jiménez et al. (2015). Both analyses use identical statistical tests (Wilcoxon signed-rank, effect size r).

````python
import re, warnings
from pathlib import Path
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from scipy.signal import correlate as _sp_correlate
import ipywidgets as widgets
from IPython.display import display

DATA_DIR        = Path(".")
_WEIGHT_EXCLUDE = {"01", "06"}
_WC_BASE_THR    = 0.5
_WC_PLAT_THR    = 11.4

_summary = pd.read_excel(DATA_DIR / "Test 6_summary.xlsx", header=2)
_summary["ID"] = _summary["ID"].astype(str).str.zfill(2)
_summary = _summary.set_index("ID")

participants = sorted([
    d.name for d in DATA_DIR.iterdir()
    if d.is_dir()
    and d.name not in {"excluded data", "exports"}
    and any(d.glob("*pressure*"))
])
print(f"Participants: {len(participants)}, excluded from weight analysis: {sorted(_WEIGHT_EXCLUDE)}")
````

````python
# ─── Helper functions (self-contained) ───────────────────────────────────────

def get_weight_direction(pid):
    if pid not in _summary.index: return ""
    return str(_summary.loc[pid, "weight direction"]).strip()

def get_condition_order(pid):
    if pid not in _summary.index: return ""
    val = str(_summary.loc[pid, "condition order"]).strip().lower()
    return "weight first" if val == "b-first" else "tempo first"

def load_imu(pid, side):
    path = DATA_DIR / pid / f"{pid}_imu_{side}.csv"
    if not path.exists(): return pd.DataFrame()
    df = pd.read_csv(path)
    if df.empty or "device_ms" not in df.columns: return df
    t = df["device_ms"].to_numpy(dtype=np.int64)
    diffs = np.diff(t)
    bad = np.where((diffs < 0) | (diffs > 2_000_000_000))[0]
    if len(bad):
        starts = np.concatenate([[0], bad + 1])
        ends   = np.concatenate([bad + 1, [len(t)]])
        best   = int(np.argmax(ends - starts))
        df     = df.iloc[starts[best]:ends[best]].reset_index(drop=True)
        t      = t[starts[best]:ends[best]]
    df["time_s"] = (t - t[0]) / 1000.0
    return df

def parse_weight_file(path):
    path = Path(path)
    if not path.exists(): return pd.DataFrame(), np.array([])
    rows_p, wc_list = [], []
    try:
        with open(path, encoding="utf-8", errors="ignore") as fh:
            for line in fh:
                line = line.strip()
                if not line: continue
                tokens = line.rstrip(";").rstrip().split()
                if len(tokens) < 3: continue
                try: wc = float(tokens[-1])
                except ValueError: continue
                unsigned_nums = re.findall(r"[\d]+\.?[\d]*", line)
                if len(unsigned_nums) < 6: continue
                try:
                    ts_ms = float(unsigned_nums[1])
                    sensors = [float(n) for n in unsigned_nums[1:] if float(n) <= 1023]
                    if len(sensors) < 4: continue
                    rows_p.append({"time_s": ts_ms/1000.0,
                                   "L_toe": sensors[0], "L_heel": sensors[1],
                                   "R_toe": sensors[2], "R_heel": sensors[3]})
                    wc_list.append(wc)
                except (ValueError, IndexError): continue
    except OSError: return pd.DataFrame(), np.array([])
    if not rows_p: return pd.DataFrame(), np.array([])
    df = pd.DataFrame(rows_p)
    df["time_s"] -= df["time_s"].iloc[0]
    return df, np.array(wc_list)

def detect_contacts(signal, t_s, baseline_mask, wc):
    if baseline_mask.sum() < 5: return [], np.nan, np.nan, np.nan
    med = float(np.median(signal[baseline_mask]))
    rng = float(np.max(signal[baseline_mask])) - med
    if rng <= 0: return [], np.nan, np.nan, np.nan
    enter = med + 0.5 * rng; exit_ = med + 0.15 * rng
    events, in_c, ts_start, i0 = [], False, None, None
    for i in range(len(signal)):
        v = signal[i]
        if not in_c and v >= enter:
            in_c, ts_start, i0 = True, t_s[i], i
        elif in_c and v <= exit_:
            in_c = False; events.append((ts_start, t_s[i], i0, i))
    contacts = []; _trapz = getattr(np, "trapezoid", np.trapz)
    for ts, te, j0, j1 in events:
        dur_ms = (te - ts) * 1000
        if not (50 < dur_ms < 2000): continue
        seg = np.maximum(signal[j0:j1+1].astype(float) - med, 0)
        impulse = float(_trapz(seg, t_s[j0:j1+1]))
        peak = float(np.max(signal[j0:j1+1]) - med)
        t_mid = ts + (te - ts) / 2
        idx = int(np.searchsorted(t_s, t_mid))
        wc_val = float(wc[idx]) if 0 <= idx < len(wc) else np.nan
        contacts.append({"t_start": ts, "t_end": te, "t_mid": t_mid,
                         "duration_ms": dur_ms, "impulse": impulse, "peak": peak, "wc": wc_val})
    return contacts, enter, exit_, med

def detect_steps_with_dwell(time_s, signal, lo_window_ms=400, hi_window_ms=3000,
                             on_ratio=0.70, off_ratio=0.35, min_range=80, debounce_ms=300):
    signal = np.asarray(signal, dtype=float); n = len(signal)
    if n == 0: return np.array([], dtype=int), np.array([], dtype=int)
    dt_ms = float(np.median(np.diff(time_s)) * 1000) if n > 1 else 5.0
    lo_win = max(8, int(round(lo_window_ms/dt_ms)))
    hi_win = max(20, int(round(hi_window_ms/dt_ms)))
    debounce = max(1, int(round(debounce_ms/dt_ms)))
    triggers, rearms = [], []; is_active, since_last = False, 999
    for i in range(n):
        since_last += 1; val = signal[i]
        lo_slice = signal[max(0, i+1-lo_win):i+1]; hi_slice = signal[max(0, i+1-hi_win):i+1]
        lo, hi = lo_slice.min(), hi_slice.max(); rng = hi - lo
        if rng > min_range and len(lo_slice) >= 8:
            on_thresh = lo + rng * on_ratio; off_thresh = lo + rng * off_ratio
            if not is_active:
                if val > on_thresh and since_last > debounce:
                    is_active, since_last = True, 0; triggers.append(i)
            elif val < off_thresh:
                is_active = False; rearms.append(i)
    if len(triggers) > len(rearms): rearms.append(n - 1)
    return np.array(triggers, dtype=int), np.array(rearms, dtype=int)

def _norm(x):
    sd = float(x.std()); return (x - x.mean()) / sd if sd > 1e-6 else np.zeros_like(x)

def sync_imu_to_weight_phase(pid):
    wf_path = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
    if not wf_path.exists(): return None, None, None
    df_p, wc = parse_weight_file(wf_path)
    if df_p.empty or len(wc) == 0: return None, None, None
    t_p = df_p["time_s"].to_numpy(); l_heel_p = df_p["L_heel"].to_numpy().astype(float)
    weight_dur = float(t_p[-1])
    df_imu_L = load_imu(pid, "L")
    if df_imu_L.empty or "heel" not in df_imu_L.columns: return None, None, None
    t_imu = df_imu_L["time_s"].to_numpy(); heel_imu = df_imu_L["heel"].to_numpy().astype(float)
    dt_imu = float(np.median(np.diff(t_imu)))
    if dt_imu <= 0: return None, None, None
    n_rs = max(2, int(weight_dur / dt_imu))
    if n_rs > len(heel_imu): return None, None, None
    t_rs = np.linspace(0, weight_dur, n_rs)
    heel_p_rs = np.interp(t_rs, t_p, l_heel_p)
    corr = _sp_correlate(_norm(heel_imu), _norm(heel_p_rs), mode="valid")
    best_lag = int(np.argmax(corr)); t_start = float(t_imu[best_lag]); t_end = t_start + weight_dur
    mask_L = (t_imu >= t_start) & (t_imu <= t_end)
    df_L_w = df_imu_L[mask_L].copy().reset_index(drop=True); df_L_w["time_s"] -= t_start
    df_imu_R = load_imu(pid, "R"); df_R_w = None
    if not df_imu_R.empty and "heel" in df_imu_R.columns:
        t_r = df_imu_R["time_s"].to_numpy(); mk_r = (t_r >= t_start) & (t_r <= t_end)
        df_R_w = df_imu_R[mk_r].copy().reset_index(drop=True); df_R_w["time_s"] -= t_start
    def wc_fn(t_local): return np.interp(t_local, t_p, wc, left=0.0, right=0.0)
    return df_L_w, df_R_w, wc_fn

def _wc_segment(wc_val):
    if abs(wc_val) <= _WC_BASE_THR: return "baseline"
    if abs(wc_val) >= _WC_PLAT_THR: return "plateau"
    return "ramp"

print("Helpers loaded.")
````

## Per-step extraction

Following Tajadura-Jiménez et al. (2015): for each step/swing, extract gait metrics and tag with the wc modulator value at that moment. Classify steps as:
- **Baseline** (`|wc| ≤ 0.5`): analog of the paper's *Control* condition
- **Plateau** (`|wc| ≥ 11.4`): analog of the paper's *Low/High frequency* condition
- **Ramp** (in between): excluded from statistical comparisons

**Pressure metrics** (FSR analog):
- `heel_contact_ms` — heel contact duration (primary metric in paper, p=0.004)
- `peak_heel_force` — peak pressure above baseline
- `gait_cycle_ms` — time between successive ipsilateral heel strikes

**IMU metrics** (accelerometer analog):
- `net_accel_peak` — max √(ax²+ay²+az²) during swing phase (proxy for paper's upward acceleration)

````python
def extract_pressure_steps(pid):
    direction = get_weight_direction(pid)
    if not direction or direction.lower() == "nan": return pd.DataFrame()
    order = get_condition_order(pid)
    path = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
    df, wc = parse_weight_file(path)
    if df.empty or len(wc) == 0: return pd.DataFrame()
    t_s = df["time_s"].to_numpy(); baseline_mask = np.abs(wc) <= _WC_BASE_THR
    rows = []
    for foot, col in [("L", "L_heel"), ("R", "R_heel")]:
        contacts, _, _, _ = detect_contacts(df[col].to_numpy(), t_s, baseline_mask, wc)
        for k, c in enumerate(contacts):
            seg = _wc_segment(c["wc"])
            if seg == "ramp": continue
            gc_ms = np.nan
            if k + 1 < len(contacts):
                gc = (contacts[k+1]["t_start"] - c["t_start"]) * 1000.0
                gc_ms = gc if 200 < gc < 3000 else np.nan
            rows.append({"pid": pid, "direction": direction, "order": order, "foot": foot,
                         "wc_segment": seg, "heel_contact_ms": c["duration_ms"],
                         "peak_heel_force": c["peak"], "gait_cycle_ms": gc_ms})
    return pd.DataFrame(rows)

def extract_imu_swings(pid):
    direction = get_weight_direction(pid)
    if not direction or direction.lower() == "nan": return pd.DataFrame()
    order = get_condition_order(pid)
    imu_L, imu_R, wc_fn = sync_imu_to_weight_phase(pid)
    rows = []
    for foot, imu_df in [("L", imu_L), ("R", imu_R)]:
        if imu_df is None or imu_df.empty: continue
        if not all(c in imu_df.columns for c in ("ax", "ay", "az", "heel")): continue
        t = imu_df["time_s"].to_numpy(); heel = imu_df["heel"].to_numpy().astype(float)
        ax = imu_df["ax"].to_numpy().astype(float)
        ay = imu_df["ay"].to_numpy().astype(float)
        az = imu_df["az"].to_numpy().astype(float)
        net = np.sqrt(ax**2 + ay**2 + az**2)
        triggers, rearms = detect_steps_with_dwell(t, heel)
        if len(triggers) < 2: continue
        for i in range(len(triggers) - 1):
            t0 = t[rearms[i]] if i < len(rearms) else t[triggers[i]]
            t1 = t[triggers[i+1]]
            mask = (t >= t0) & (t <= t1)
            if mask.sum() < 3: continue
            wc_mid = float(wc_fn(np.array([(t0+t1)/2]))[0])
            seg = _wc_segment(wc_mid)
            if seg == "ramp": continue
            net_swing = net[mask]
            rows.append({"pid": pid, "direction": direction, "order": order, "foot": foot,
                         "wc_segment": seg, "net_accel_peak": float(net_swing.max()),
                         "net_accel_rms": float(np.sqrt(np.mean(net_swing**2))),
                         "swing_duration_ms": (t1 - t0) * 1000.0})
    return pd.DataFrame(rows)

print("Extraction functions defined.")
````

## Per-participant time series explorer

Select a participant to see their full weight session time series.
The **wc modulator** is shown as background shading — grey = baseline (`|wc| ≤ 0.5`),
green = plateau (`|wc| ≥ 11.4`), white = ramp transition.

- **Top panel**: Heel contact duration (ms) per step — L foot (blue), R foot (red)
- **Bottom panel**: Swing-phase peak net acceleration √(ax²+ay²+az²) per swing — from synced IMU

The orange line in each panel is a rolling mean (window adjustable with the slider).
IMU data requires cross-correlation sync; if sync fails for a participant only the pressure panel appears.

````python
import ipywidgets as widgets
from IPython.display import display

# Participants available for the explorer (exclude bad IMU ones)
_ts_pids = sorted([
    p for p in participants
    if p not in _WEIGHT_EXCLUDE
    and (DATA_DIR / p / f"{p}_pressure_weight.csv").exists()
    and get_weight_direction(p) not in ("", "nan")
])

_ts_pid_dd  = widgets.Dropdown(options=_ts_pids,
                                value=_ts_pids[0] if _ts_pids else None,
                                description="Participant:",
                                style={"description_width": "90px"},
                                layout=widgets.Layout(width="200px"))
_ts_ma_sl   = widgets.IntSlider(min=1, max=20, step=1, value=5,
                                 description="Roll avg (steps):",
                                 style={"description_width": "115px"},
                                 layout=widgets.Layout(width="320px"))
_ts_out     = widgets.Output()


def _shade_wc(ax, t_s, wc, alpha_base=0.12, alpha_plat=0.12):
    """Draw baseline (grey) and plateau (green) background bands."""
    in_base = in_plat = False
    t_start = None
    for i, (t, w) in enumerate(zip(t_s, wc)):
        seg_now = _wc_segment(w)
        if seg_now == "baseline" and not in_base:
            in_base = True; t_start = t
        elif seg_now != "baseline" and in_base:
            ax.axvspan(t_start, t, color="#aaaaaa", alpha=alpha_base, lw=0)
            in_base = False
        if seg_now == "plateau" and not in_plat:
            in_plat = True; t_start = t
        elif seg_now != "plateau" and in_plat:
            ax.axvspan(t_start, t, color="#2ecc71", alpha=alpha_plat, lw=0)
            in_plat = False
    # Close any open spans at end
    if in_base:
        ax.axvspan(t_start, t_s[-1], color="#aaaaaa", alpha=alpha_base, lw=0)
    if in_plat:
        ax.axvspan(t_start, t_s[-1], color="#2ecc71", alpha=alpha_plat, lw=0)


def _rolling_mean(x, window):
    """Simple centred rolling mean, clipping edges."""
    out = np.full_like(x, np.nan, dtype=float)
    for i in range(len(x)):
        lo = max(0, i - window // 2)
        hi = min(len(x), i + window // 2 + 1)
        vals = x[lo:hi]
        valid = vals[~np.isnan(vals)]
        if len(valid) > 0:
            out[i] = float(np.mean(valid))
    return out


def plot_ts(pid, ma_window):
    direction = get_weight_direction(pid)

    # ── Load pressure data ────────────────────────────────────────────
    path = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
    df_p, wc = parse_weight_file(path)
    if df_p.empty:
        print(f"P{pid}: no pressure weight data"); return

    t_wc = df_p["time_s"].to_numpy()
    baseline_mask = np.abs(wc) <= _WC_BASE_THR

    # Extract contacts for L and R
    contacts_by_foot = {}
    for foot, col in [("L", "L_heel"), ("R", "R_heel")]:
        contacts, _, _, _ = detect_contacts(df_p[col].to_numpy(), t_wc, baseline_mask, wc)
        contacts_by_foot[foot] = contacts

    # ── Sync IMU ──────────────────────────────────────────────────────
    imu_L, imu_R, wc_fn = sync_imu_to_weight_phase(pid)
    has_imu = (imu_L is not None and not imu_L.empty
               and all(c in imu_L.columns for c in ("ax", "ay", "az", "heel")))

    n_rows = 2 if has_imu else 1
    fig, axes = plt.subplots(n_rows, 1,
                             figsize=(14, 4.5 * n_rows),
                             sharex=False)
    if n_rows == 1:
        axes = [axes]
    fig.subplots_adjust(hspace=0.45)

    # ── Panel 1: Heel contact duration ────────────────────────────────
    ax1 = axes[0]
    _shade_wc(ax1, t_wc, wc)

    FOOT_COLORS = {"L": "#1f77b4", "R": "#d62728"}
    for foot, contacts in contacts_by_foot.items():
        if not contacts:
            continue
        t_arr  = np.array([c["t_mid"]       for c in contacts])
        dur_arr= np.array([c["duration_ms"]  for c in contacts])
        color  = FOOT_COLORS[foot]
        ax1.scatter(t_arr, dur_arr, color=color, s=14, alpha=0.55,
                    label=f"{foot} foot", zorder=3)
        rm = _rolling_mean(dur_arr, ma_window)
        ax1.plot(t_arr, rm, color=color, lw=1.8, alpha=0.9, zorder=4)

    ax1.set_ylabel("Heel contact duration (ms)", fontsize=9)
    ax1.set_xlabel("Time (s)", fontsize=9)
    ax1.legend(fontsize=8, loc="upper right")
    ax1.grid(linestyle="--", alpha=0.3)

    # ── Expected behaviour overlay — heel contact ─────────────────────
    # Paper: heavier sound (increasing wc) → LONGER heel contact
    #        lighter sound (decreasing wc) → SHORTER heel contact
    # wc sign already matches expected direction (both positive for increasing,
    # both negative for decreasing), so overlay wc / 12 scaled to data range.
    _all_dur = np.array([c["duration_ms"]
                         for cs in contacts_by_foot.values() for c in cs])
    if len(_all_dur) > 4:
        _y_mid = float(np.median(_all_dur))
        _y_amp = float(np.percentile(_all_dur, 75) - np.percentile(_all_dur, 25)) * 0.7
        _exp1  = _y_mid + (wc / 12.0) * _y_amp
        _elbl1 = ("Expected ↑ contact (heavier sound)" if direction == "increasing"
                  else "Expected ↓ contact (lighter sound)")
        ax1.plot(t_wc, _exp1, color="#e74c3c", lw=1.4, ls="--", alpha=0.65,
                 label=_elbl1, zorder=2)

    # Annotation patches for legend
    from matplotlib.patches import Patch
    handles, labels = ax1.get_legend_handles_labels()
    handles += [Patch(color="#aaaaaa", alpha=0.4, label="Baseline wc"),
                Patch(color="#2ecc71", alpha=0.4, label="Plateau wc")]
    ax1.legend(handles=handles, fontsize=7, loc="upper right")
    ax1.set_title(f"P{pid}  ({direction})  —  Heel contact duration  "
                  f"[rolling avg window = {ma_window} steps]",
                  fontsize=10)

    # ── Panel 2: Swing net acceleration (IMU) ────────────────────────
    if has_imu:
        ax2 = axes[1]

        for foot, imu_df in [("L", imu_L), ("R", imu_R)]:
            if imu_df is None or imu_df.empty: continue
            if not all(c in imu_df.columns for c in ("ax","ay","az","heel")): continue

            t_i    = imu_df["time_s"].to_numpy()
            heel_i = imu_df["heel"].to_numpy().astype(float)
            net_i  = np.sqrt(imu_df["ax"].to_numpy()**2 +
                             imu_df["ay"].to_numpy()**2 +
                             imu_df["az"].to_numpy()**2).astype(float)

            triggers, rearms = detect_steps_with_dwell(t_i, heel_i)
            if len(triggers) < 2: continue

            sw_t, sw_v = [], []
            for k in range(len(triggers) - 1):
                t0   = t_i[rearms[k]] if k < len(rearms) else t_i[triggers[k]]
                t1   = t_i[triggers[k + 1]]
                mask = (t_i >= t0) & (t_i <= t1)
                if mask.sum() < 3: continue
                sw_t.append((t0 + t1) / 2)
                sw_v.append(float(net_i[mask].max()))

            if not sw_t: continue
            sw_t_arr = np.array(sw_t); sw_v_arr = np.array(sw_v)
            color = FOOT_COLORS[foot]
            ax2.scatter(sw_t_arr, sw_v_arr, color=color, s=14, alpha=0.55,
                        label=f"{foot} foot", zorder=3)
            rm2 = _rolling_mean(sw_v_arr, ma_window)
            ax2.plot(sw_t_arr, rm2, color=color, lw=1.8, alpha=0.9, zorder=4)

        # Shade using wc on IMU time axis (approximate)
        imu_t = imu_L["time_s"].to_numpy()
        wc_on_imu = wc_fn(imu_t)
        _shade_wc(ax2, imu_t, wc_on_imu)

        ax2.set_ylabel("Swing net acceleration peak\n√(ax²+ay²+az²)", fontsize=9)
        ax2.set_xlabel("Time (s, weight-phase relative)", fontsize=9)
        handles2, _ = ax2.get_legend_handles_labels()
        handles2 += [Patch(color="#aaaaaa", alpha=0.4, label="Baseline wc"),
                     Patch(color="#2ecc71", alpha=0.4, label="Plateau wc")]
        ax2.legend(handles=handles2, fontsize=7, loc="upper right")
        ax2.grid(linestyle="--", alpha=0.3)

        # ── Expected behaviour overlay — swing acceleration ───────────────
        # Paper: heavier sound (increasing wc) → LOWER swing acceleration
        #        lighter sound (decreasing wc) → HIGHER swing acceleration
        # Direction is OPPOSITE to wc, so multiply wc_norm by -1.
        # Scale using the plotted y-range so no redundant computation needed.
        _sw_lo, _sw_hi = ax2.get_ylim()
        _sy_mid = (_sw_lo + _sw_hi) / 2
        _sy_amp = (_sw_hi - _sw_lo) * 0.35
        _exp2   = _sy_mid - (wc_on_imu / 12.0) * _sy_amp   # minus = opposite direction
        _elbl2  = ("Expected ↓ swing (heavier sound)" if direction == "increasing"
                   else "Expected ↑ swing (lighter sound)")
        ax2.plot(imu_t, _exp2, color="#e74c3c", lw=1.4, ls="--", alpha=0.65,
                 label=_elbl2, zorder=2)

        ax2.set_title(f"P{pid}  —  Swing-phase peak net acceleration  "
                      f"(IMU, synced via cross-correlation)",
                      fontsize=10)

    plt.tight_layout()
    return fig


def _refresh_ts(_=None):
    import io as _io
    _ts_out.clear_output(wait=True)
    with _ts_out:
        _fig = plot_ts(_ts_pid_dd.value, _ts_ma_sl.value)
        if _fig is not None:
            _buf = _io.BytesIO()
            _fig.savefig(_buf, dpi=150, bbox_inches='tight', format='png')
            _buf.seek(0)
            display(_fig); plt.close(_fig)
            _btn = widgets.Button(description='Export', button_style='info',
                                   icon='download',
                                   layout=widgets.Layout(width='110px'))
            _lbl = widgets.Label('')
            def _do_save(_, buf=_buf, pid=_ts_pid_dd.value):
                _lbl.value = 'Saving…'
                d = DATA_DIR / 'exports' / 'paper_replication'
                d.mkdir(parents=True, exist_ok=True)
                buf.seek(0)
                (d / f'timeseries_{pid}.png').write_bytes(buf.read())
                _lbl.value = f'✓  timeseries_{pid}.png'
            _btn.on_click(_do_save)
            display(widgets.HBox([_btn, _lbl]))

_ts_pid_dd.observe(_refresh_ts, names="value")
_ts_ma_sl.observe(_refresh_ts,  names="value")
_refresh_ts()

display(widgets.HBox([_ts_pid_dd, _ts_ma_sl]), _ts_out)
````

## Per-session coupling grid — heel contact & swing acceleration

Recreates the cadence-coupling grid from `pressure_timeseries.ipynb` but for the
weight condition, using the two biomechanical gait metrics instead of cadence:

- **Top grid**: heel contact duration (ms) per step — from pressure insoles
- **Bottom grid**: swing net acceleration √(ax²+ay²+az²) per swing — from synced IMU

Each panel shows one participant's full weight session.
Grey shading = baseline wc (`|wc| ≤ 0.5`), green = plateau (`|wc| ≥ 11.4`).
The **black regression line** is fitted over the active manipulation window
(`|wc| > 0.5`), regressing the metric against the wc modulator value.
**Title colour**: green = slope in the expected direction (per Tajadura-Jiménez 2015),
red = opposite direction.

Expected directions:
- Increasing (heavier): heel contact **↑**, swing accel **↓**  → positive slope for heel, negative for swing
- Decreasing (lighter): heel contact **↓**, swing accel **↑**  → negative slope for heel, positive for swing

Sessions are sorted by coupling strength (absolute slope, descending).

````python
from scipy.stats import linregress as _lr_fn

_WGRID_INC = "#e67e22"   # orange — increasing direction
_WGRID_DEC = "#9b59b6"   # purple — decreasing direction


def _build_weight_sessions():
    """
    For each weight session extract per-step heel contact duration and
    per-swing net acceleration, fit linear regression of metric vs wc
    in the active window, and sort by |slope|.
    """
    sessions = []
    for pid in participants:
        if pid in _WEIGHT_EXCLUDE:
            continue
        direction = get_weight_direction(pid)
        if not direction or direction.lower() == "nan":
            continue

        path = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
        df_p, wc = parse_weight_file(path)
        if df_p.empty or len(wc) == 0:
            continue

        t_s           = df_p["time_s"].to_numpy()
        baseline_mask = np.abs(wc) <= _WC_BASE_THR

        # ── Heel contact duration per step (L + R pooled) ──────────────
        all_contacts = []
        for col in ("L_heel", "R_heel"):
            cs, _, _, _ = detect_contacts(df_p[col].to_numpy(), t_s, baseline_mask, wc)
            all_contacts.extend(cs)
        all_contacts.sort(key=lambda c: c["t_mid"])

        t_heel  = np.array([c["t_mid"]       for c in all_contacts])
        dur_heel = np.array([c["duration_ms"] for c in all_contacts])
        wc_heel = np.interp(t_heel, t_s, wc) if len(t_heel) > 0 else np.array([])
        act_heel = np.abs(wc_heel) > _WC_BASE_THR

        slope_heel = np.nan
        if len(t_heel) >= 5:
            fit = _lr_fn(t_heel, dur_heel)
            slope_heel = float(fit.slope)

        # ── Swing net acceleration per swing (from synced IMU) ──────────
        imu_L, imu_R, wc_fn = sync_imu_to_weight_phase(pid)
        t_sw = np.array([]); acc_sw = np.array([]); wc_sw = np.array([])
        slope_sw = np.nan

        for imu_df in [imu_L, imu_R]:
            if imu_df is None or imu_df.empty: continue
            if not all(c in imu_df.columns for c in ("ax","ay","az","heel")): continue
            t_i    = imu_df["time_s"].to_numpy()
            heel_i = imu_df["heel"].to_numpy().astype(float)
            net_i  = np.sqrt(imu_df["ax"].to_numpy()**2 +
                             imu_df["ay"].to_numpy()**2 +
                             imu_df["az"].to_numpy()**2).astype(float)
            trigs, rearms = detect_steps_with_dwell(t_i, heel_i)
            for k in range(len(trigs) - 1):
                t0 = t_i[rearms[k]] if k < len(rearms) else t_i[trigs[k]]
                t1 = t_i[trigs[k + 1]]
                mk = (t_i >= t0) & (t_i <= t1)
                if mk.sum() < 3: continue
                t_sw  = np.append(t_sw,  (t0 + t1) / 2)
                acc_sw = np.append(acc_sw, float(net_i[mk].max()))
                wc_sw  = np.append(wc_sw,  float(wc_fn(np.array([(t0+t1)/2]))[0]))

        act_sw = np.abs(wc_sw) > _WC_BASE_THR
        if len(t_sw) >= 5:
            fit = _lr_fn(t_sw, acc_sw)
            slope_sw = float(fit.slope)

        sessions.append({
            "pid":       pid, "direction": direction,
            "t_wc": t_s, "wc": wc,
            "t_heel": t_heel, "dur_heel": dur_heel,
            "wc_heel": wc_heel, "act_heel": act_heel, "slope_heel": slope_heel,
            "t_sw": t_sw,   "acc_sw": acc_sw,
            "wc_sw": wc_sw, "act_sw": act_sw,   "slope_sw": slope_sw,
        })

    # Sort by abs(slope_heel) descending, NaN last
    sessions.sort(key=lambda s: -abs(s["slope_heel"]) if not np.isnan(s["slope_heel"]) else -999)
    return sessions


def _draw_weight_grid(sessions, metric="heel", save_path=None):
    """
    Draw per-session coupling grid.  metric = "heel" or "swing".
    Black regression line fitted over active wc window.
    Title colour: green = slope in expected direction, red = opposite.
    """
    if not sessions:
        print("No sessions to plot."); return None

    ncols = 5
    nrows = int(np.ceil(len(sessions) / ncols))
    fig, axes = plt.subplots(nrows, ncols,
                             figsize=(3.3 * ncols, 2.6 * nrows),
                             squeeze=False)
    for ax in axes.ravel():
        ax.axis("off")

    for ax, d in zip(axes.ravel(), sessions):
        ax.axis("on")
        color = _WGRID_INC if d["direction"] == "increasing" else _WGRID_DEC

        if metric == "heel":
            t, y, wc_m, act = d["t_heel"], d["dur_heel"], d["wc_heel"], d["act_heel"]
            slope  = d["slope_heel"]
            ylabel = "Heel contact (ms)"
            # Expected: increasing → positive slope; decreasing → negative slope
            expected_pos = (d["direction"] == "increasing")
        else:
            t, y, wc_m, act = d["t_sw"], d["acc_sw"], d["wc_sw"], d["act_sw"]
            slope  = d["slope_sw"]
            ylabel = "Swing net accel"
            # Expected: increasing → negative slope (heavier → less dynamic)
            #           decreasing → positive slope (lighter → more dynamic)
            expected_pos = (d["direction"] == "decreasing")

        if len(t) == 0:
            ax.text(0.5, 0.5, "No IMU data", ha="center", va="center",
                    transform=ax.transAxes, color="gray", fontsize=8)
            ax.set_title(f"P{d['pid']}", fontsize=8.5); continue

        # Background shading (grey = baseline, green = plateau)
        _shade_wc(ax, d["t_wc"], d["wc"])

        # Scatter: active = colored, inactive = grey
        if act.sum() > 0:
            ax.scatter(t[act],  y[act],  s=6, alpha=0.35, color=color, edgecolors="none")
            ax.scatter(t[~act], y[~act], s=6, alpha=0.12, color="gray", edgecolors="none")
        else:
            ax.scatter(t, y, s=6, alpha=0.30, color=color, edgecolors="none")

        # Rolling mean
        rm = _rolling_mean(y, 5)
        ax.plot(t, rm, color=color, lw=1.4, alpha=0.85)

        # Straight regression line fitted against time (y = a + b*t)
        if len(t) >= 5:
            fit_t = _lr_fn(t, y)
            ax.plot(t, fit_t.intercept + fit_t.slope * t,
                    color="black", lw=1.8, alpha=0.75)

        # Title: green if slope in expected direction
        good = (slope > 0) == expected_pos if not np.isnan(slope) else None
        title_col = "#2a7a2a" if good else ("#c44" if good is not None else "gray")
        slope_str = f"{slope:+.2f}" if not np.isnan(slope) else "n/a"
        exp_arrow = "↑" if expected_pos == (slope > 0) else "↑" if expected_pos else "↓"
        exp_label = ("Expected ↑" if expected_pos else "Expected ↓")
        ax.set_title(
            f"P{d['pid']}  {d['direction'][:4]}  s={slope_str}  [{exp_label}]",
            fontsize=7.5, color=title_col
        )
        ax.set_xlabel("time (s)", fontsize=7)
        ax.set_ylabel(ylabel, fontsize=7)
        ax.tick_params(labelsize=6.5)
        ax.grid(alpha=0.25)

    lbl = "Heel Contact Duration" if metric == "heel" else "Swing Net Acceleration"
    fig.suptitle(
        f"Per-session {lbl} vs weight manipulation\n"
        "grey=baseline wc  green=plateau wc  "
        "black line=linear fit vs time (gradient = overall trend)  "
        "green title=gradient in expected direction",
        fontsize=9,
    )
    fig.tight_layout(rect=(0, 0, 1, 0.95))
    if save_path:
        fig.savefig(save_path, dpi=150, bbox_inches="tight")
    return fig


print("Building weight coupling sessions (IMU sync included — ~2 min)…")
_wgt_sessions = _build_weight_sessions()
print(f"Sessions: {len(_wgt_sessions)}")

_n_heel_pos  = sum(1 for s in _wgt_sessions
                    if not np.isnan(s["slope_heel"])
                    and (s["slope_heel"] > 0) == (s["direction"] == "increasing"))
_n_swing_pos = sum(1 for s in _wgt_sessions
                    if not np.isnan(s["slope_sw"])
                    and (s["slope_sw"] > 0) == (s["direction"] == "decreasing"))
print(f"Heel slope in expected direction:  {_n_heel_pos}/{len(_wgt_sessions)}")
print(f"Swing slope in expected direction: {_n_swing_pos}/{len(_wgt_sessions)}")

# ── Draw & display both grids ─────────────────────────────────────────────────
import io as _io

for _metric, _label in [("heel", "Heel contact duration"), ("swing", "Swing net acceleration")]:
    _fig = _draw_weight_grid(_wgt_sessions, metric=_metric)
    if _fig is None: continue
    _buf = _io.BytesIO()
    _fig.savefig(_buf, dpi=150, bbox_inches="tight", format="png")
    _buf.seek(0)
    display(_fig); plt.close(_fig)

    _btn = widgets.Button(description=f"Export {_label[:12]}",
                           button_style="info", icon="download",
                           layout=widgets.Layout(width="170px"))
    _lbl_w = widgets.Label("")
    def _mk_save(buf=_buf, metric=_metric):
        def _save(_):
            _lbl_w.value = "Saving…"
            d = DATA_DIR / "exports" / "paper_replication"
            d.mkdir(parents=True, exist_ok=True)
            buf.seek(0)
            (d / f"wgt_coupling_{metric}.png").write_bytes(buf.read())
            _lbl_w.value = f"✓  wgt_coupling_{metric}.png"
        return _save
    _btn.on_click(_mk_save())
    display(widgets.HBox([_btn, _lbl_w]))
````

````python
# ─── Statistical significance of per-session coupling slopes ──────────────────
# Tests whether the regression gradients (time vs metric) are consistently
# in the expected direction across participants.
#
# Expected signs:
#   Heel contact:     increasing → slope > 0 (longer contact)
#                     decreasing → slope < 0 (shorter contact)
#   Swing accel:      increasing → slope < 0 (less dynamic)
#                     decreasing → slope > 0 (more dynamic)
#
# Method: sign-correct each slope to the expected direction, then run a
# one-sample Wilcoxon signed-rank test against zero.

from scipy.stats import wilcoxon as _wlcx
import numpy as np, pandas as pd

_ALPHA = 0.05

def _coupling_stats(sessions, metric, expected_sign_fn):
    """
    expected_sign_fn(direction) → +1 or -1 depending on expected slope direction.
    Returns dict of results.
    """
    slopes = []
    for s in sessions:
        raw = s[f"slope_{metric}"]
        if np.isnan(raw):
            continue
        expected = expected_sign_fn(s["direction"])
        slopes.append({"raw": raw, "corrected": raw * expected,
                        "in_expected": (raw * expected) > 0,
                        "direction": s["direction"]})
    if len(slopes) < 3:
        return None
    corrected = np.array([s["corrected"] for s in slopes])
    n = len(corrected)
    n_exp = sum(s["in_expected"] for s in slopes)
    try:
        _, p_wlcx = _wlcx(corrected, zero_method="zsplit")
    except Exception:
        p_wlcx = np.nan
    z = np.nan if np.isnan(p_wlcx) else abs(float(
        __import__("scipy").stats.norm.ppf(p_wlcx / 2)))
    r = z / np.sqrt(n) if not np.isnan(z) else np.nan
    sig = ("***" if p_wlcx < 0.001 else "**" if p_wlcx < 0.01
           else "*" if p_wlcx < _ALPHA else "ns")
    return {
        "n": n, "n_expected": n_exp,
        "pct_expected": 100 * n_exp / n,
        "median_corrected_slope": float(np.median(corrected)),
        "p_wilcoxon": float(p_wlcx),
        "r": float(r),
        "sig": sig,
    }

print("=== Per-session coupling slope significance ===")
print(f"(One-sample Wilcoxon signed-rank vs 0, sign-corrected to expected direction)")
print(f"α = {_ALPHA}\n")

for metric, label, exp_fn in [
    ("heel",
     "Heel contact duration",
     lambda d: +1 if d == "increasing" else -1),
    ("sw",
     "Swing acceleration",
     lambda d: -1 if d == "increasing" else +1),
]:
    res = _coupling_stats(_wgt_sessions, metric, exp_fn)
    if res is None:
        print(f"{label}: insufficient data"); continue
    print(f"{label}")
    print(f"  n = {res['n']}  |  in expected direction: "
          f"{res['n_expected']}/{res['n']} ({res['pct_expected']:.0f}%)")
    print(f"  median corrected slope: {res['median_corrected_slope']:+.4f}")
    print(f"  Wilcoxon vs 0:  p = {res['p_wilcoxon']:.4f}  {res['sig']}  "
          f"r = {res['r']:.3f}")
    print()


# ── Null check: B1-baseline-only drift to validate the linear regression ──────
print("\n=== Null check: slope during B1 baseline (pre-manipulation) ===")
print("If B1-only slope matches expected direction → natural drift confound")
print("(Expected: ~50% by chance if no confound)\n")

_b1_results = {"heel": 0, "swing": 0, "n": 0}
for pid in participants:
    if pid in _WEIGHT_EXCLUDE: continue
    path = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
    if not path.exists(): continue
    df_p, wc = parse_weight_file(path)
    if df_p.empty: continue
    t_s = df_p["time_s"].to_numpy()
    b1_mask = (np.abs(wc) <= 0.5) & (t_s < t_s[-1] * 0.5)
    if b1_mask.sum() < 15: continue
    direction = get_weight_direction(pid)
    if not direction or direction.lower() == "nan": continue
    exp_sign = +1 if direction == "increasing" else -1
    baseline_mask_full = np.abs(wc) <= 0.5
    heel_b1 = []
    for col in ("L_heel", "R_heel"):
        cs, _, _, _ = detect_contacts(df_p[col].to_numpy(), t_s, baseline_mask_full, wc)
        heel_b1.extend([c["duration_ms"] for c in cs if b1_mask[np.searchsorted(t_s, c["t_mid"])]])
    if len(heel_b1) >= 5:
        t_b1 = np.linspace(0, 1, len(heel_b1))
        from scipy.stats import linregress as _lrb
        slope_b1, *_ = _lrb(t_b1, heel_b1)
        if (slope_b1 * exp_sign) > 0:
            _b1_results["heel"] += 1
        _b1_results["n"] += 1

n_b1 = _b1_results["n"]
if n_b1 > 0:
    heel_pct = 100 * _b1_results["heel"] / n_b1
    print(f"Heel contact B1-baseline expected direction: "
          f"{_b1_results['heel']}/{n_b1} ({heel_pct:.0f}%)")
    if heel_pct > 60:
        print("  → HIGH baseline drift rate: linear regression confounded by natural gait settling.")
        print("  → Linear regression slopes are NOT reliable evidence of manipulation effects.")
        print("  → Use Pearson r (wc target correlation) as the primary statistical test.")
    else:
        print("  → Baseline drift rate near chance: linear regression may be informative.")
````

## Per-session achievement plots — heel contact & swing acceleration

Mirrors the `plot_tempo_achievement` style from `pressure_timeseries.ipynb`:

- **y-axis**: % change from individual baseline (mean during `|wc| ≤ 0.5` segments)
- **Reference line** (dashed): the `weight_change` modulator scaled to ±8% to show the expected direction and magnitude of effect — same ±8% reference used in the tempo analysis
- **Left panel**: heel contact duration — expected ↑ for increasing (heavier), ↓ for decreasing (lighter)
- **Right panel**: swing-phase peak net acceleration — expected ↓ for increasing, ↑ for decreasing

The stats box shows baseline, plateau mean, % change achieved, and whether the direction matches the prediction.

````python
def _contact_pct_timeseries(all_contacts, baseline_dur):
    """Convert contact list to (t_array, pct_change_array)."""
    if not all_contacts or baseline_dur < 1e-6:
        return np.array([]), np.array([])
    t_arr   = np.array([c["t_mid"]      for c in all_contacts])
    dur_arr = np.array([c["duration_ms"] for c in all_contacts])
    return t_arr, (dur_arr / baseline_dur - 1.0) * 100.0


def _step_rolling_mean(t, y, window_steps):
    """Rolling mean over ±window_steps/2 contacts (step-based, not time-based)."""
    half = max(1, window_steps // 2)
    out  = np.array([float(np.mean(y[max(0, i-half):i+half+1]))
                     for i in range(len(y))])
    return t, out


def _swing_pct_timeseries(imu_df, wc_fn, baseline_gy):
    """
    Extract per-swing peak net acceleration, expressed as % change from baseline.
    Returns (t_array, pct_change_array).
    """
    if imu_df is None or imu_df.empty or baseline_gy < 1e-6:
        return np.array([]), np.array([])
    if not all(c in imu_df.columns for c in ("ax", "ay", "az", "heel")):
        return np.array([]), np.array([])
    t    = imu_df["time_s"].to_numpy()
    heel = imu_df["heel"].to_numpy().astype(float)
    net  = np.sqrt(imu_df["ax"].to_numpy()**2 +
                   imu_df["ay"].to_numpy()**2 +
                   imu_df["az"].to_numpy()**2).astype(float)
    triggers, rearms = detect_steps_with_dwell(t, heel)
    if len(triggers) < 2:
        return np.array([]), np.array([])
    sw_t, sw_pct = [], []
    for k in range(len(triggers) - 1):
        t0   = t[rearms[k]] if k < len(rearms) else t[triggers[k]]
        t1   = t[triggers[k + 1]]
        mask = (t >= t0) & (t <= t1)
        if mask.sum() < 3:
            continue
        peak = float(net[mask].max())
        sw_t.append((t0 + t1) / 2)
        sw_pct.append((peak / baseline_gy - 1.0) * 100.0)
    return np.array(sw_t), np.array(sw_pct)


def plot_weight_achievement(pid, window_steps=7):
    """
    Two-panel per-session plot mirroring plot_tempo_achievement style.
    Panel 1: Heel contact duration % change from baseline
    Panel 2: Swing-phase peak net acceleration % change from baseline
    Reference line: wc modulator scaled to ±8% showing expected effect direction.
    """
    direction = get_weight_direction(pid)
    if not direction or direction.lower() == "nan":
        return None

    # ── Load pressure + contacts ─────────────────────────────────────────────
    path = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
    df_p, wc = parse_weight_file(path)
    if df_p.empty:
        return None
    t_wc          = df_p["time_s"].to_numpy()
    baseline_mask = np.abs(wc) <= _WC_BASE_THR

    all_contacts = []
    for col in ("L_heel", "R_heel"):
        cs, _, _, _ = detect_contacts(df_p[col].to_numpy(), t_wc, baseline_mask, wc)
        all_contacts.extend(cs)
    all_contacts.sort(key=lambda c: c["t_mid"])
    if not all_contacts:
        return None

    base_c    = [c for c in all_contacts if abs(c["wc"]) <= _WC_BASE_THR]
    plat_c    = [c for c in all_contacts if abs(c["wc"]) >= _WC_PLAT_THR]
    if len(base_c) < 3:
        return None

    baseline_dur = float(np.mean([c["duration_ms"] for c in base_c]))
    plateau_dur  = float(np.mean([c["duration_ms"] for c in plat_c])) if plat_c else np.nan
    dur_pct_plat = (plateau_dur / baseline_dur - 1.0) * 100.0 if not np.isnan(plateau_dur) else np.nan

    t_dur, pct_dur   = _contact_pct_timeseries(all_contacts, baseline_dur)
    t_dur, dur_ma    = _step_rolling_mean(t_dur, pct_dur, window_steps)

    # ── Sync IMU + swing ─────────────────────────────────────────────────────
    imu_L, imu_R, wc_fn = sync_imu_to_weight_phase(pid)
    has_imu = (imu_L is not None and not imu_L.empty)

    baseline_gy = np.nan
    t_sw = pct_sw = np.array([])
    if has_imu:
        t_i   = imu_L["time_s"].to_numpy()
        wc_i  = wc_fn(t_i)
        bm_i  = np.abs(wc_i) <= _WC_BASE_THR
        if "ax" in imu_L.columns and bm_i.sum() > 5:
            net_i = np.sqrt(imu_L["ax"]**2 + imu_L["ay"]**2 + imu_L["az"]**2).to_numpy()
            heel_i = imu_L["heel"].to_numpy().astype(float)
            tr, re = detect_steps_with_dwell(t_i, heel_i)
            base_sw = []
            for k in range(len(tr) - 1):
                t0 = t_i[re[k]] if k < len(re) else t_i[tr[k]]
                t1 = t_i[tr[k+1]]
                mask = (t_i >= t0) & (t_i <= t1)
                if mask.sum() < 3: continue
                wc_mid = float(wc_fn(np.array([(t0+t1)/2]))[0])
                if abs(wc_mid) <= _WC_BASE_THR:
                    base_sw.append(float(net_i[mask].max()))
            if base_sw:
                baseline_gy = float(np.mean(base_sw))
        t_sw, pct_sw = _swing_pct_timeseries(imu_L, wc_fn, baseline_gy)
        # Rolling mean for swing
        if len(t_sw) > window_steps:
            _, pct_sw_ma = _step_rolling_mean(t_sw, pct_sw, window_steps)
        else:
            pct_sw_ma = pct_sw

    # Swing plateau % change
    if has_imu and len(t_sw) > 0:
        plat_sw = [pct_sw[k] for k, tt in enumerate(t_sw)
                   if abs(wc_fn(np.array([tt]))[0]) >= _WC_PLAT_THR]
        sw_pct_plat = float(np.mean(plat_sw)) if plat_sw else np.nan
    else:
        sw_pct_plat = np.nan

    # ── Reference line: wc modulator scaled to ±8% (expected direction) ──────
    # Heel: same direction as wc (heavier → longer → positive)
    # Swing: opposite to wc (heavier → less dynamic → negative)
    ref_dur   =  (wc / 12.0) * 8.0   # ±8% reference for heel contact
    ref_swing = -(wc / 12.0) * 8.0   # ±8% reference for swing accel

    # ── Figure ───────────────────────────────────────────────────────────────
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    fig.subplots_adjust(wspace=0.35)

    order = get_condition_order(pid)

    def _draw_panel(ax, t_pts, pct_raw, pct_ma, ref_t, ref_line,
                    baseline_val, plat_pct, metric_label, units, ref_label, color):
        ax.axhline(0, color="#bdc3c7", lw=1.0, zorder=1)
        # Reference line
        ax.plot(ref_t, ref_line, color="#2c3e50", lw=1.4, ls="--",
                label=f"Expected direction (±8% ref)", zorder=2)
        ax.fill_between(ref_t, ref_line - 2, ref_line + 2,
                        color="#2c3e50", alpha=0.07, zorder=1)
        # Background shading
        _shade_wc(ax, ref_t, wc)
        # Raw + MA
        if len(t_pts) > 0:
            ax.scatter(t_pts, pct_raw, color=color, s=10, alpha=0.3, zorder=3)
            ax.plot(t_pts, pct_ma, color=color, lw=1.8,
                    label=f"{metric_label} (rolling avg)", zorder=4)
        # Stats box
        n_match = "✓" if not np.isnan(plat_pct) and (
            (ref_line[len(ref_line)//2] > 0 and plat_pct > 0) or
            (ref_line[len(ref_line)//2] < 0 and plat_pct < 0)
        ) else "✗"
        stats_txt = (
            f"Baseline:    {baseline_val:.1f} {units}\n"
            f"% at plateau:{plat_pct:+.1f}%\n"
            f"Direction:   {n_match}"
        )
        ax.text(0.02, 0.97, stats_txt, transform=ax.transAxes, fontsize=8,
                va="top", family="monospace",
                bbox=dict(boxstyle="round,pad=0.4", facecolor="white", alpha=0.85))
        ax.set_xlabel("Time in weight session (s)", fontsize=9)
        ax.set_ylabel("% change from baseline", fontsize=9)
        ax.legend(fontsize=7, loc="lower right")
        ax.grid(ls="--", alpha=0.3)

    _draw_panel(
        ax1, t_dur, pct_dur, dur_ma,
        t_wc, ref_dur,
        baseline_dur, dur_pct_plat,
        "Heel contact duration", "ms",
        "Expected (heavier=+, lighter=−)", "#1f77b4"
    )
    ax1.set_title(f"P{pid} ({direction}) — Heel contact duration", fontsize=10)

    if has_imu and len(t_sw) > 0:
        _draw_panel(
            ax2, t_sw, pct_sw, pct_sw_ma,
            imu_L["time_s"].to_numpy(),
            -(wc_fn(imu_L["time_s"].to_numpy()) / 12.0) * 8.0,
            baseline_gy, sw_pct_plat,
            "Swing net acceleration", "g",
            "Expected (heavier=−, lighter=+)", "#d62728"
        )
        ax2.set_title(f"P{pid} ({direction}) — Swing-phase net acceleration", fontsize=10)
    else:
        ax2.text(0.5, 0.5, "IMU sync unavailable", ha="center", va="center",
                 transform=ax2.transAxes, color="gray", fontsize=12)
        ax2.set_title(f"P{pid} — Swing acceleration (no IMU)", fontsize=10)

    fig.suptitle(
        f"T6 P{pid}  |  {direction} weight  |  {order}\n"
        "% change from individual baseline  —  dashed = expected direction (±8% ref scale)",
        fontsize=10, y=1.02,
    )
    plt.tight_layout()
    return fig


# ── Widget ────────────────────────────────────────────────────────────────────
_ach_pids = sorted([
    p for p in participants
    if p not in _WEIGHT_EXCLUDE
    and (DATA_DIR / p / f"{p}_pressure_weight.csv").exists()
    and get_weight_direction(p) not in ("", "nan")
])

_ach_pid_dd = widgets.Dropdown(
    options=_ach_pids,
    value=_ach_pids[0] if _ach_pids else None,
    description="Participant:",
    style={"description_width": "90px"},
    layout=widgets.Layout(width="200px"),
)
_ach_win_sl = widgets.IntSlider(
    min=1, max=20, step=1, value=7,
    description="Roll avg (steps):",
    style={"description_width": "115px"},
    layout=widgets.Layout(width="320px"),
)
_ach_out = widgets.Output()


def _refresh_ach(_=None):
    import io as _io
    _ach_out.clear_output(wait=True)
    with _ach_out:
        _fig = plot_weight_achievement(_ach_pid_dd.value, _ach_win_sl.value)
        if _fig is not None:
            _buf = _io.BytesIO()
            _fig.savefig(_buf, dpi=150, bbox_inches="tight", format="png")
            _buf.seek(0)
            display(_fig); plt.close(_fig)
            _btn = widgets.Button(description="Export", button_style="info",
                                   icon="download",
                                   layout=widgets.Layout(width="110px"))
            _lbl = widgets.Label("")
            def _do_save(_, buf=_buf, pid=_ach_pid_dd.value):
                _lbl.value = "Saving…"
                d = DATA_DIR / "exports" / "paper_replication"
                d.mkdir(parents=True, exist_ok=True)
                buf.seek(0)
                (d / f"achievement_{pid}.png").write_bytes(buf.read())
                _lbl.value = f"✓  achievement_{pid}.png"
            _btn.on_click(_do_save)
            display(widgets.HBox([_btn, _lbl]))
        else:
            print("No data available for this participant.")


_ach_pid_dd.observe(_refresh_ach, names="value")
_ach_win_sl.observe(_refresh_ach, names="value")
_refresh_ach()
display(widgets.HBox([_ach_pid_dd, _ach_win_sl]), _ach_out)
````

## Average trajectory by group — weight condition

Mirrors the `_draw_avg_2x2` figure from `pressure_timeseries.ipynb` (tempo analysis).

Each panel shows all individual session curves (transparent) plus the group mean ± 1 SEM,
plotted on a **normalised 0–100% trial-progress axis**. The **dashed dark line** is the
normalised weight-change modulator (|wc| / 12 × direction sign, scaled to ±1) — the
"trapezoid" target shape showing what a perfect response would look like.

Groups: **direction** (increasing / decreasing) × **condition order** (weight first / tempo first).

Two 2×2 figures: one for heel contact duration, one for swing-phase acceleration.

````python
# ─── Average trajectory builder (weight condition) ────────────────────────────

COMMON_GRID_W = np.linspace(0, 1, 300)

def _build_avg_weight_trajectories(metric="heel"):
    """
    For each of the 4 groups (direction × order), compute mean ± SEM of the
    z-scored metric on a normalised 0–100% time grid.

    metric = "heel"  → heel contact duration
             "swing" → swing-phase net acceleration (requires IMU sync)

    Returns (result_dict, COMMON_GRID_W, mean_target_by_direction)
    where result_dict is keyed by group label.
    """
    groups = {}  # label → {"pids": [], "direction": "", "order": ""}
    for pid in participants:
        if pid in _WEIGHT_EXCLUDE:
            continue
        direction = get_weight_direction(pid)
        if not direction or direction.lower() == "nan":
            continue
        if not (DATA_DIR / pid / f"{pid}_pressure_weight.csv").exists():
            continue
        order = get_condition_order(pid)
        label = f"{direction.title()} — {order}"
        groups.setdefault(label, {"pids": [], "direction": direction, "order": order})
        groups[label]["pids"].append(pid)

    result = {}
    target_by_dir = {}   # direction → mean target curve

    for label, meta in groups.items():
        pids      = meta["pids"]
        direction = meta["direction"]
        ds        = +1 if direction == "increasing" else -1   # heel direction sign
        if metric == "swing":
            ds = -ds   # swing is opposite

        curves  = []
        targets = []

        for pid in pids:
            path = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
            df_p, wc = parse_weight_file(path)
            if df_p.empty or len(wc) == 0:
                continue
            t_s = df_p["time_s"].to_numpy()
            baseline_mask = np.abs(wc) <= _WC_BASE_THR

            # ── Target: normalised wc shape ──────────────────────────────
            target_raw = (np.abs(wc) / 12.0) * ds   # ±1 trapezoid
            t_n        = t_s / t_s[-1]               # normalise time 0→1
            targets.append(np.interp(COMMON_GRID_W, t_n, target_raw))

            if metric == "heel":
                # ── Heel contact duration ──────────────────────────────
                all_contacts = []
                for col in ("L_heel", "R_heel"):
                    cs, _, _, _ = detect_contacts(
                        df_p[col].to_numpy(), t_s, baseline_mask, wc)
                    all_contacts.extend(cs)
                if not all_contacts:
                    continue
                t_c   = np.array([c["t_mid"]      for c in all_contacts])
                dur_c = np.array([c["duration_ms"] for c in all_contacts], dtype=float)
                mu, sd = dur_c.mean(), dur_c.std()
                z_c = (dur_c - mu) / sd if sd > 1e-6 else np.zeros_like(dur_c)
                t_n_c = t_c / t_s[-1]
                curves.append(np.interp(COMMON_GRID_W, t_n_c, z_c))

            else:
                # ── Swing-phase net acceleration ──────────────────────
                imu_L, imu_R, wc_fn = sync_imu_to_weight_phase(pid)
                sw_t, sw_z = [], []
                for imu_df in [imu_L, imu_R]:
                    if imu_df is None or imu_df.empty: continue
                    if not all(c in imu_df.columns for c in ("ax","ay","az","heel")):
                        continue
                    t_i    = imu_df["time_s"].to_numpy()
                    heel_i = imu_df["heel"].to_numpy().astype(float)
                    net_i  = np.sqrt(imu_df["ax"].to_numpy()**2 +
                                     imu_df["ay"].to_numpy()**2 +
                                     imu_df["az"].to_numpy()**2).astype(float)
                    trigs, rearms = detect_steps_with_dwell(t_i, heel_i)
                    for k in range(len(trigs) - 1):
                        t0 = t_i[rearms[k]] if k < len(rearms) else t_i[trigs[k]]
                        t1 = t_i[trigs[k+1]]
                        mk = (t_i >= t0) & (t_i <= t1)
                        if mk.sum() < 3: continue
                        sw_t.append((t0 + t1) / 2)
                        sw_z.append(float(net_i[mk].max()))
                if len(sw_t) < 5:
                    continue
                sw_t_arr = np.array(sw_t); sw_v_arr = np.array(sw_z, dtype=float)
                mu, sd = sw_v_arr.mean(), sw_v_arr.std()
                z_sw = (sw_v_arr - mu) / sd if sd > 1e-6 else np.zeros_like(sw_v_arr)
                sw_t_n = sw_t_arr / t_s[-1]
                curves.append(np.interp(COMMON_GRID_W, sw_t_n, z_sw))

        if not curves:
            result[label] = None
            continue

        arr       = np.array(curves)
        mean_c    = arr.mean(axis=0)
        sem_c     = arr.std(axis=0) / np.sqrt(len(arr))
        final_z   = float(np.median(arr[:, -20:].mean(axis=1)))
        mean_tgt  = np.array(targets).mean(axis=0) if targets else None

        result[label] = {
            "mean": mean_c, "sem": sem_c, "curves": list(arr),
            "n": len(arr), "direction": direction, "order": meta["order"],
            "target_sign": ds, "final_z": final_z,
            "mean_target": mean_tgt, "label": label,
        }
        if direction not in target_by_dir and mean_tgt is not None:
            target_by_dir[direction] = mean_tgt

    return result, COMMON_GRID_W, target_by_dir


def _draw_weight_avg_2x2(metric="heel", save_path=None):
    """
    Render the 2×2 average-trajectory figure for the weight condition.
    metric = "heel" or "swing"
    """
    data, t_common, target_by_dir = _build_avg_weight_trajectories(metric)

    ROWS = ["increasing", "decreasing"]
    COLS = ["weight first", "tempo first"]
    DIR_COLOR = {"increasing": "#e67e22", "decreasing": "#9b59b6"}

    metric_label = "Heel contact duration (z-score)" if metric == "heel" else                    "Swing net acceleration (z-score)"
    expected_desc = {
        "increasing": "↑ heel / ↓ swing",
        "decreasing": "↓ heel / ↑ swing",
    }

    fig, axes = plt.subplots(2, 2, figsize=(14, 8), sharey=False, sharex=True)
    fig.subplots_adjust(hspace=0.42, wspace=0.28)

    t_pct = t_common * 100   # 0–100%

    # Phase bands — approximate weight session structure on normalised axis
    # B1: 0-16%, Ramp↑: 16-23%, Plateau: 23-74%, Ramp↓: 74-81%, B2: 81-100%
    PHASE_BANDS_W = [
        (0,   16, "#aaaaaa", 0.06),   # Baseline 1
        (16,  23, "#3498db", 0.07),   # Ascending ramp
        (23,  74, "#e67e22", 0.05),   # Plateau
        (74,  81, "#27ae60", 0.07),   # Descending ramp
        (81, 100, "#aaaaaa", 0.06),   # Baseline 2
    ]

    for ri, direction in enumerate(ROWS):
        for ci, order in enumerate(COLS):
            ax = axes[ri][ci]
            color = DIR_COLOR[direction]

            key = next(
                (k for k, v in data.items()
                 if v is not None
                 and v["direction"] == direction
                 and order in k.lower()),
                None,
            )

            # Phase shading
            for x0, x1, c, alpha in PHASE_BANDS_W:
                ax.axvspan(x0, x1, alpha=alpha, color=c)
            ax.axhline(0, color="#bdc3c7", lw=0.8)

            # Target trapezoid from mean wc of this direction
            tgt = target_by_dir.get(direction)
            if tgt is None and key and data[key]:
                tgt = data[key]["mean_target"]
            if tgt is not None:
                ax.plot(t_pct, tgt, color="#2c3e50", lw=1.4, ls="--", alpha=0.8,
                        label="Target (|wc|/12 × sign, ±1)")
                ax.fill_between(t_pct, tgt - 0.08, tgt + 0.08,
                                color="#2c3e50", alpha=0.06)

            if key is None or data[key] is None:
                ax.text(0.5, 0.5, "No data", ha="center", va="center",
                        transform=ax.transAxes, color="gray", fontsize=11)
                ax.set_title(f"{direction.title()} — {order}", fontsize=10)
                continue

            d = data[key]
            # Individual curves
            for curve in d["curves"]:
                ax.plot(t_pct, curve, color=color, alpha=0.18, lw=0.9, zorder=2)
            # Mean ± SEM
            ax.plot(t_pct, d["mean"], color=color, lw=2.2, zorder=4,
                    label=f"Mean (n={d['n']})")
            ax.fill_between(t_pct,
                            d["mean"] - d["sem"],
                            d["mean"] + d["sem"],
                            color=color, alpha=0.18, zorder=3)

            ax.text(0.97, 0.97,
                    "n = " + str(d["n"]) + "\nMedian final: " + f"{d['final_z']:+.2f}" + " z",
                    transform=ax.transAxes, fontsize=8, va="top", ha="right",
                    bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))
            ax.set_title(f"{direction.title()} — {order}  "
                         f"({expected_desc[direction]})",
                         fontsize=10, color=color)
            ax.set_xlim(0, 100)
            ax.set_ylim(-1.8, 1.8)
            ax.grid(ls="--", alpha=0.3)
            if ri == 1:
                ax.set_xlabel("Trial progress (%)", fontsize=9)
            if ci == 0:
                ax.set_ylabel(metric_label, fontsize=9)
            if ri == 0 and ci == 0:
                ax.legend(fontsize=7, loc="lower right")

    fig.suptitle(
        "T6 — Average " + metric_label + " by group (weight condition)\n"
        "(shaded band = +/-1 SEM; dashed = normalised wc target +/-1; "
        "phases: B1 | Ramp up | Plateau | Ramp down | B2)",
        fontsize=10, y=1.01,
    )
    plt.tight_layout()
    if save_path:
        fig.savefig(save_path, dpi=150, bbox_inches="tight")
    return fig


# ── Render both grids ─────────────────────────────────────────────────────────
print("Building average trajectories (swing requires IMU sync — ~2 min)…")
_fig_heel = _draw_weight_avg_2x2(metric="heel")
plt.show()
_fig_swing = _draw_weight_avg_2x2(metric="swing")
plt.show()

# ── Export buttons ────────────────────────────────────────────────────────────
import io as _io_avg

_avg_figs = [(_fig_heel, "avg_2x2_heel"), (_fig_swing, "avg_2x2_swing")]
_avg_exp_btn = widgets.Button(description="Export 2×2 figures", button_style="info",
                               icon="download", layout=widgets.Layout(width="170px"))
_avg_exp_lbl = widgets.Label("")

def _do_avg_exp(_):
    _avg_exp_lbl.value = "Saving…"
    d = DATA_DIR / "exports" / "paper_replication"
    d.mkdir(parents=True, exist_ok=True)
    for fig, fname in _avg_figs:
        fig.savefig(d / f"{fname}.png", dpi=150, bbox_inches="tight")
    _avg_exp_lbl.value = "✓ Saved to exports/paper_replication/"

_avg_exp_btn.on_click(_do_avg_exp)
display(widgets.HBox([_avg_exp_btn, _avg_exp_lbl]))
````

## Ramp vs plateau — directional bias analysis

Every participant's weight session follows a fixed 5-phase modulator structure:

```
Baseline₁ (~20 s) → Ascending ramp (~9 s) → Plateau (~60 s) → Descending ramp (~9 s) → Baseline₂
    |wc| ≈ 0              0 → ±12               |wc| ≈ 12           ±12 → 0               |wc| ≈ 0
```

The baseline-vs-plateau test (previous section) captures whether gait changes once the
manipulation is fully established. But some participants may respond more strongly to the
**onset of change** — the 9-second ascending ramp when the audio is actively accelerating
toward its target level — than to the sustained plateau. This would indicate a
**directional bias**: sensitivity to the *direction* of change rather than its absolute level.

**Metrics defined for each participant:**
- `plateau_response` = mean z-score(heel contact) at plateau − mean at Baseline₁
- `ramp_response` = mean z-score(heel contact) at ascending ramp − mean at Baseline₁
- `bias_index` = ramp_response − plateau_response
  - **> 0** → more affected during ramp (onset-sensitive / directional bias)
  - **≈ 0** → equal response at ramp and plateau
  - **< 0** → more affected at plateau (sustained-level response)

The interactive explorer below shows each participant's step-level trajectory with phase
boundaries marked, and a stimulus-response curve coloured by phase.

````python
# ─── Phase classification ──────────────────────────────────────────────────────

def classify_phases(wc, t_s):
    """
    Identify the 5 temporal phases of the weight session from the wc modulator.
    Returns dict: phase_name → (t_start, t_end) in seconds.
    Ascending ramp  = |wc| increasing away from 0.
    Descending ramp = |wc| decreasing back toward 0.
    """
    abs_wc = np.abs(wc)
    n      = len(wc)

    # Label every sample: 'B' baseline, 'U' ascending ramp, 'P' plateau, 'D' descending ramp
    labels = []
    for i, w in enumerate(wc):
        aw = abs(w)
        if aw <= _WC_BASE_THR:
            labels.append("B")
        elif aw >= _WC_PLAT_THR:
            labels.append("P")
        else:
            # ramp — ascending if |wc| increasing, descending if decreasing
            if i == 0 or abs_wc[i] >= abs_wc[i - 1]:
                labels.append("U")
            else:
                labels.append("D")

    # Collect contiguous runs
    runs = []
    cur_lbl = labels[0]; cur_start = 0
    for i in range(1, n):
        if labels[i] != cur_lbl:
            runs.append((cur_lbl, t_s[cur_start], t_s[i - 1]))
            cur_lbl = labels[i]; cur_start = i
    runs.append((cur_lbl, t_s[cur_start], t_s[-1]))

    # Map first/second occurrence of B and U/D to named phases
    phase_map = {"baseline_1": None, "ramp_up": None, "plateau": None,
                 "ramp_down": None, "baseline_2": None}
    b_count = u_count = d_count = 0
    for lbl, ts, te in runs:
        if lbl == "B":
            b_count += 1
            if b_count == 1:   phase_map["baseline_1"] = (ts, te)
            elif b_count == 2: phase_map["baseline_2"] = (ts, te)
        elif lbl == "U":
            u_count += 1
            if u_count == 1:   phase_map["ramp_up"] = (ts, te)
        elif lbl == "D":
            d_count += 1
            if d_count == 1:   phase_map["ramp_down"] = (ts, te)
        elif lbl == "P":
            if phase_map["plateau"] is None:
                phase_map["plateau"] = (ts, te)

    return phase_map


def _phase_of(t_mid, phases):
    """Return the phase name that contains time t_mid, or None."""
    for name, interval in phases.items():
        if interval is not None and interval[0] <= t_mid <= interval[1]:
            return name
    return None


def compute_phase_z(pid):
    """
    Compute per-phase z-scored peak heel force (primary metric) for one participant.

    Key statistical features (ramp_response, plateau_response, bias_index)
    use **manipulation-window thirds**: the period from when |wc| first exceeds
    0.5 to when it returns below 0.5 is split into three equal time thirds:
      - Early third: ascending ramp + early plateau (onset response)
      - Mid third:   sustained plateau core (sustained response)
      - Late third:  late plateau + descending ramp (fade/correction)

    This gives ~42-54 contacts per phase (vs ~12-15 for the exact 9s ramp)
    and is robust to individual differences in response timing — it does not
    require the gait response to be temporally aligned with the wc modulator.

    The exact wc-threshold phases (B1/Ramp+/Plateau/Ramp-/B2) are also
    returned for bar-chart visualisations only.
    """
    direction = get_weight_direction(pid)
    if not direction or direction.lower() == "nan":
        return None
    order = get_condition_order(pid)

    path = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
    df_p, wc = parse_weight_file(path)
    if df_p.empty or len(wc) == 0:
        return None

    t_s           = df_p["time_s"].to_numpy()
    baseline_mask = np.abs(wc) <= _WC_BASE_THR
    phases        = classify_phases(wc, t_s)

    # Collect all contacts (L + R)
    all_contacts = []
    for col in ("L_heel", "R_heel"):
        contacts, _, _, _ = detect_contacts(df_p[col].to_numpy(), t_s, baseline_mask, wc)
        all_contacts.extend(contacts)

    if len(all_contacts) < 10:
        return None

    import math

    # ── Z-score BOTH duration and peak force across all contacts ─────────
    # Peak heel force is the primary metric (significant in group analysis,
    # p=0.039, r=0.73 for increasing direction).
    # Duration is retained as a secondary metric for bar chart comparison.
    def _zscore(arr):
        mu, sd = arr.mean(), arr.std()
        return (arr - mu) / sd if sd > 1e-6 else np.zeros_like(arr)

    all_dur   = np.array([c["duration_ms"] for c in all_contacts])
    all_force = np.array([c["peak"]        for c in all_contacts])
    if all_dur.std() < 1e-6 or all_force.std() < 1e-6:
        return None

    z_dur   = _zscore(all_dur)
    z_force = _zscore(all_force)

    # ── Exact wc-threshold phases (for bar chart — both metrics) ─────────
    def mean_or_nan(lst):
        return float(np.mean(lst)) if len(lst) >= 2 else np.nan

    ph_force = {ph: [] for ph in phases}
    ph_dur   = {ph: [] for ph in phases}
    for c, zf, zd in zip(all_contacts, z_force, z_dur):
        ph = _phase_of(c["t_mid"], phases)
        if ph is not None:
            ph_force[ph].append(zf)
            ph_dur[ph].append(zd)

    z_b1       = mean_or_nan(ph_force["baseline_1"])
    z_up       = mean_or_nan(ph_force["ramp_up"])
    z_plat     = mean_or_nan(ph_force["plateau"])
    z_down     = mean_or_nan(ph_force["ramp_down"])
    z_b2       = mean_or_nan(ph_force["baseline_2"])
    z_b1_dur   = mean_or_nan(ph_dur["baseline_1"])
    z_up_dur   = mean_or_nan(ph_dur["ramp_up"])
    z_plat_dur = mean_or_nan(ph_dur["plateau"])
    z_down_dur = mean_or_nan(ph_dur["ramp_down"])
    z_b2_dur   = mean_or_nan(ph_dur["baseline_2"])

    # ── Manipulation-window thirds — PRIMARY classification (peak force) ──
    manip_mask = np.abs(wc) > 0.5
    manip_t = t_s[manip_mask]
    if len(manip_t) < 10:
        return None

    t_ms = float(manip_t[0]); t_me = float(manip_t[-1])
    md   = t_me - t_ms
    t_e1 = t_ms + md / 3.0; t_e2 = t_ms + 2.0 * md / 3.0

    thirds_f = {"baseline": [], "early": [], "mid": [], "late": []}
    thirds_d = {"baseline": [], "early": [], "mid": [], "late": []}
    for c, zf, zd in zip(all_contacts, z_force, z_dur):
        tm = c["t_mid"]
        if tm < t_ms:       bucket = "baseline"
        elif tm <= t_e1:    bucket = "early"
        elif tm <= t_e2:    bucket = "mid"
        elif tm <= t_me:    bucket = "late"
        else:               continue
        thirds_f[bucket].append(zf)
        thirds_d[bucket].append(zd)

    # Force-based features (primary — drives classification)
    z_ref_f   = mean_or_nan(thirds_f["baseline"])
    z_early_f = mean_or_nan(thirds_f["early"])
    z_mid_f   = mean_or_nan(thirds_f["mid"])
    z_late_f  = mean_or_nan(thirds_f["late"])
    ref_f     = z_ref_f if (not math.isnan(z_ref_f) and len(thirds_f["baseline"]) >= 3) else 0.0
    ramp_resp  = (z_early_f - ref_f) if not math.isnan(z_early_f) else np.nan
    plat_resp  = (z_mid_f   - ref_f) if not math.isnan(z_mid_f)   else np.nan
    bias_index = ((ramp_resp - plat_resp)
                  if not math.isnan(ramp_resp) and not math.isnan(plat_resp)
                  else np.nan)

    # Duration-based features (secondary — for comparison)
    z_ref_d   = mean_or_nan(thirds_d["baseline"])
    z_early_d = mean_or_nan(thirds_d["early"])
    z_mid_d   = mean_or_nan(thirds_d["mid"])
    z_late_d  = mean_or_nan(thirds_d["late"])
    ref_d     = z_ref_d if (not math.isnan(z_ref_d) and len(thirds_d["baseline"]) >= 3) else 0.0
    ramp_resp_dur = (z_early_d - ref_d) if not math.isnan(z_early_d) else np.nan
    plat_resp_dur = (z_mid_d   - ref_d) if not math.isnan(z_mid_d)   else np.nan

    # ── Pearson r with normalised wc target (supplementary) ──────────────
    from scipy.ndimage import gaussian_filter1d as _gf1d
    t_grid   = np.linspace(0, float(t_s[-1]), 200)
    t_c      = np.array([c["t_mid"] for c in all_contacts])
    zf_interp = np.interp(t_grid, t_c, z_force)
    zf_sm     = _gf1d(zf_interp.astype(float), sigma=4)
    ds         = +1 if direction == "increasing" else -1
    target_grid = np.interp(t_grid, t_s, (np.abs(wc) / 12.0) * ds)
    if np.std(zf_sm) > 1e-6 and np.std(target_grid) > 1e-6:
        influence_r = float(np.corrcoef(zf_sm, target_grid)[0, 1])
    else:
        influence_r = 0.0

    return {
        "pid": pid, "direction": direction, "order": order,
        # Exact wc-threshold phases — force (primary, for bar chart)
        "z_base1": z_b1, "z_ramp_up": z_up, "z_plateau": z_plat,
        "z_ramp_down": z_down, "z_base2": z_b2,
        # Exact wc-threshold phases — duration (secondary, for bar chart)
        "z_base1_dur": z_b1_dur, "z_ramp_up_dur": z_up_dur,
        "z_plateau_dur": z_plat_dur, "z_ramp_down_dur": z_down_dur,
        "z_base2_dur": z_b2_dur,
        # Manipulation-window thirds — force (primary → drives classification)
        "z_early": z_early_f, "z_mid": z_mid_f, "z_late": z_late_f,
        # Manipulation-window thirds — duration (secondary)
        "z_early_dur": z_early_d, "z_mid_dur": z_mid_d, "z_late_dur": z_late_d,
        # Classification features (from peak force)
        "ramp_response":     ramp_resp,
        "plateau_response":  plat_resp,
        "bias_index":        bias_index,
        "ramp_response_dur": ramp_resp_dur,
        "plateau_response_dur": plat_resp_dur,
        "influence_r":       influence_r,
        "phases":            phases,
    }


def build_phase_stats():
    """Compute compute_phase_z for all valid participants. Returns DataFrame."""
    rows = []
    for pid in participants:
        if pid in _WEIGHT_EXCLUDE:
            continue
        r = compute_phase_z(pid)
        if r is not None:
            rows.append({k: v for k, v in r.items() if k != "phases"})
    return pd.DataFrame(rows)


# ─── Swing-phase equivalent of compute_phase_z ────────────────────────────────

def compute_phase_z_swing(pid):
    """
    Same as compute_phase_z() but uses per-swing net IMU acceleration instead
    of heel contact duration.  Direction sign is FLIPPED for swing:
      increasing (heavier sound) → swing accel expected to DECREASE  (ds_swing = -1)
      decreasing (lighter sound) → swing accel expected to INCREASE  (ds_swing = +1)
    """
    direction = get_weight_direction(pid)
    if not direction or direction.lower() == "nan":
        return None
    order = get_condition_order(pid)

    # Need pressure file to get wc modulator + timing
    path_p = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
    df_p, wc = parse_weight_file(path_p)
    if df_p.empty or len(wc) == 0:
        return None
    t_s = df_p["time_s"].to_numpy()

    # Sync IMU and extract per-swing net acceleration
    imu_L, imu_R, wc_fn = sync_imu_to_weight_phase(pid)
    all_swings = []
    for imu_df in [imu_L, imu_R]:
        if imu_df is None or imu_df.empty: continue
        if not all(c in imu_df.columns for c in ("ax","ay","az","heel")): continue
        t_i    = imu_df["time_s"].to_numpy()
        heel_i = imu_df["heel"].to_numpy().astype(float)
        net_i  = np.sqrt(imu_df["ax"].to_numpy()**2 +
                         imu_df["ay"].to_numpy()**2 +
                         imu_df["az"].to_numpy()**2).astype(float)
        trigs, rearms = detect_steps_with_dwell(t_i, heel_i)
        for k in range(len(trigs) - 1):
            t0 = t_i[rearms[k]] if k < len(rearms) else t_i[trigs[k]]
            t1 = t_i[trigs[k+1]]
            mk = (t_i >= t0) & (t_i <= t1)
            if mk.sum() < 3: continue
            all_swings.append({"t_mid": (t0+t1)/2, "acc": float(net_i[mk].max())})

    if len(all_swings) < 10:
        return None

    # Z-score across all swings in session
    all_acc = np.array([s["acc"] for s in all_swings])
    mu, sd  = all_acc.mean(), all_acc.std()
    if sd < 1e-6:
        return None
    z_all = (all_acc - mu) / sd

    # Exact wc-threshold phases (for bar chart)
    phases   = classify_phases(wc, t_s)
    phase_sw = {ph: [] for ph in phases}
    for s, z in zip(all_swings, z_all):
        ph = _phase_of(s["t_mid"], phases)
        if ph is not None:
            phase_sw[ph].append(z)

    def mean_or_nan(lst):
        return float(np.mean(lst)) if len(lst) >= 2 else np.nan

    z_b1   = mean_or_nan(phase_sw["baseline_1"])
    z_up   = mean_or_nan(phase_sw["ramp_up"])
    z_plat = mean_or_nan(phase_sw["plateau"])
    z_down = mean_or_nan(phase_sw["ramp_down"])
    z_b2   = mean_or_nan(phase_sw["baseline_2"])

    # Manipulation-window thirds
    import math
    manip_mask = np.abs(wc) > 0.5
    manip_t = t_s[manip_mask]
    if len(manip_t) < 10:
        return None
    t_ms   = float(manip_t[0]); t_me = float(manip_t[-1])
    md     = t_me - t_ms
    t_e1   = t_ms + md / 3.0; t_e2 = t_ms + 2.0 * md / 3.0

    thirds_sw = {"baseline": [], "early": [], "mid": [], "late": []}
    for s, z in zip(all_swings, z_all):
        tm = s["t_mid"]
        if tm < t_ms:             thirds_sw["baseline"].append(z)
        elif tm <= t_e1:          thirds_sw["early"].append(z)
        elif tm <= t_e2:          thirds_sw["mid"].append(z)
        elif tm <= t_me:          thirds_sw["late"].append(z)

    z_ref   = mean_or_nan(thirds_sw["baseline"])
    z_early = mean_or_nan(thirds_sw["early"])
    z_mid   = mean_or_nan(thirds_sw["mid"])
    z_late  = mean_or_nan(thirds_sw["late"])
    ref_val = z_ref if (not math.isnan(z_ref) and len(thirds_sw["baseline"]) >= 3) else 0.0

    ramp_resp  = (z_early - ref_val) if not math.isnan(z_early) else np.nan
    plat_resp  = (z_mid   - ref_val) if not math.isnan(z_mid)   else np.nan
    bias_index = ((ramp_resp - plat_resp)
                  if not math.isnan(ramp_resp) and not math.isnan(plat_resp)
                  else np.nan)

    # Pearson r — note swing direction is FLIPPED vs heel
    from scipy.ndimage import gaussian_filter1d as _gf1d
    t_sw   = np.array([s["t_mid"] for s in all_swings])
    t_grid = np.linspace(0, float(t_s[-1]), 200)
    z_interp = np.interp(t_grid, t_sw, z_all)
    z_sm     = _gf1d(z_interp.astype(float), sigma=4)
    ds_swing = -1 if direction == "increasing" else +1  # FLIPPED
    target_grid = np.interp(t_grid, t_s, (np.abs(wc) / 12.0) * ds_swing)
    if np.std(z_sm) > 1e-6 and np.std(target_grid) > 1e-6:
        influence_r = float(np.corrcoef(z_sm, target_grid)[0, 1])
    else:
        influence_r = 0.0

    return {
        "pid": pid, "direction": direction, "order": order,
        "z_base1": z_b1, "z_ramp_up": z_up, "z_plateau": z_plat,
        "z_ramp_down": z_down, "z_base2": z_b2,
        "z_early": z_early, "z_mid": z_mid, "z_late": z_late,
        "ramp_response":    ramp_resp,
        "plateau_response": plat_resp,
        "bias_index":       bias_index,
        "influence_r":      influence_r,
        "phases":           phases,
    }


def build_phase_stats_swing():
    """Compute compute_phase_z_swing for all valid participants."""
    rows = []
    for pid in participants:
        if pid in _WEIGHT_EXCLUDE:
            continue
        r = compute_phase_z_swing(pid)
        if r is not None:
            rows.append({k: v for k, v in r.items() if k != "phases"})
    return pd.DataFrame(rows)


print("Phase functions defined. Building group stats…")
df_phase = build_phase_stats()
print(f"Phase stats: {len(df_phase)} participants")
print(df_phase[["pid","direction","z_base1","z_ramp_up","z_plateau",
                "ramp_response","plateau_response","bias_index"]].round(3).to_string(index=False))


# ─── Interactive per-participant plot ─────────────────────────────────────────

_pb_pid_dd = widgets.Dropdown(
    options=_ts_pids,
    value=_ts_pids[0] if _ts_pids else None,
    description="Participant:",
    style={"description_width": "90px"},
    layout=widgets.Layout(width="200px"),
)
_pb_ma_sl = widgets.IntSlider(
    min=1, max=20, step=1, value=5,
    description="Roll avg (steps):",
    style={"description_width": "115px"},
    layout=widgets.Layout(width="320px"),
)
_pb_out = widgets.Output()

_PHASE_COLORS = {
    "baseline_1": "#aaaaaa", "ramp_up": "#e67e22",
    "plateau": "#2ecc71",    "ramp_down": "#3498db",
    "baseline_2": "#aaaaaa",
}
_PHASE_LABELS = {
    "baseline_1": "B₁",  "ramp_up": "Ramp↑",
    "plateau": "Plateau", "ramp_down": "Ramp↓",
    "baseline_2": "B₂",
}


from scipy.ndimage import gaussian_filter1d as _gf1d

def _classify_behaviour(r):
    """
    Classify a weight session into 4 behavioural categories.
    All features are direction-corrected (positive = expected-direction response).

    Classification order matters:
      1. Subconscious correction checked first (negative ramp must be caught
         before consistent-influence check, which would otherwise misclassify it)
      2. Drift-phase only: clear onset response that fades to plateau
      3. Consistent influence: sustained plateau response
      4. No clear effect: default

    Thresholds calibrated to the actual data distribution (responses are small
    in this within-session continuous design — much smaller than the paper's
    between-conditions paradigm).
    """
    ds = +1 if r["direction"] == "increasing" else -1
    correct_ramp  = r["ramp_response"]   * ds   # direction-corrected onset response
    correct_plat  = r["plateau_response"] * ds   # direction-corrected plateau response
    plateau_drop  = correct_ramp - correct_plat  # positive = onset faded at plateau

    # 1. Wrong-direction onset → subconscious correction (ramp went opposite)
    if correct_ramp < 0:
        if correct_plat > 0.05:
            return "Subconscious correction",       "#3498db"
        return "No clear effect",               "#bdc3c7"

    # 2. Onset response that fades at plateau → drift-phase sensitivity
    if correct_ramp > 0.04 and plateau_drop > 0.04:
        return "Impact only from drift phases", "#e67e22"

    # 3. Sustained plateau response → consistent influence
    if correct_plat > 0.10:
        return "Consistent influence",          "#2ecc71"

    return "No clear effect",               "#bdc3c7"


def _items_to_smooth_z(items, val_key, t_grid, sigma=4):
    """
    Interpolate per-step/swing items onto t_grid, z-score vs all values,
    Gaussian-smooth and return the array.
    """
    if not items:
        return np.zeros(len(t_grid))
    t_pts = np.array([x["t"]        for x in items])
    vals  = np.array([x[val_key]    for x in items], dtype=float)
    mu, sd = vals.mean(), vals.std()
    z_raw = (vals - mu) / sd if sd > 1e-6 else np.zeros_like(vals)
    z_interp = np.interp(t_grid, t_pts, z_raw,
                         left=float(z_raw[0]) if len(z_raw) else 0.0,
                         right=float(z_raw[-1]) if len(z_raw) else 0.0)
    return _gf1d(z_interp.astype(float), sigma=sigma)


def _draw_metric_panel(ax, items, val_key, t_wc, wc,
                       direction_sign, metric_label, phases, t_grid=None):
    """
    Draw one metric panel (heel contact or swing accel) with:
    - Phase-coloured raw scatter
    - Smoothed z-score line (dark)
    - Normalized wc trapezoid target (dashed)
    - Green fill where metric is in expected direction, red where opposite
    - Phase boundary verticals + mean horizontal lines
    - Pearson r in title annotation
    """
    phase_order = ["baseline_1","ramp_up","plateau","ramp_down","baseline_2"]

    # Build regular time grid
    if t_grid is None:
        t_grid = np.linspace(float(t_wc[0]), float(t_wc[-1]), 300)

    # Smoothed z-score on grid
    z_smooth = _items_to_smooth_z(items, val_key, t_grid)

    # Normalised target trapezoid: |wc| / 12 × direction_sign
    target = np.interp(t_grid, t_wc, (np.abs(wc) / 12.0) * direction_sign)

    # Pearson r between smoothed metric and target
    if np.std(z_smooth) > 1e-6 and np.std(target) > 1e-6:
        r_val = float(np.corrcoef(z_smooth, target)[0, 1])
    else:
        r_val = np.nan

    # ── Target trapezoid (dashed) ─────────────────────────────────────────
    ax.plot(t_grid, target, color="#2c3e50", lw=1.6, ls="--", alpha=0.85,
            label=f"Target (wc normalised  ±1)", zorder=5)
    ax.fill_between(t_grid, target - 0.05, target + 0.05,
                    color="#2c3e50", alpha=0.07, zorder=1)

    # ── Raw scatter coloured by phase ─────────────────────────────────────
    if items:
        all_vals = np.array([x[val_key] for x in items], dtype=float)
        mu, sd   = all_vals.mean(), all_vals.std()
        for x in items:
            ph  = x.get("phase")
            z_v = (x[val_key] - mu) / sd if sd > 1e-6 else 0.0
            col = _PHASE_COLORS.get(ph, "#cccccc") if ph else "#cccccc"
            ax.scatter(x["t"], z_v, color=col, s=12, alpha=0.45, zorder=3)

    # ── Smoothed metric line ──────────────────────────────────────────────
    ax.plot(t_grid, z_smooth, color="#222", lw=1.8, alpha=0.9, zorder=4,
            label=f"Metric (smoothed)  r={r_val:+.2f}" if not np.isnan(r_val)
                  else "Metric (smoothed)")

    # ── Green / red fill (expected vs opposing direction) ─────────────────
    # Green where z_smooth × direction_sign > 0 (moving in expected direction)
    # Red where < 0
    in_dir  = (z_smooth * direction_sign) > 0
    opp_dir = (z_smooth * direction_sign) < 0
    ax.fill_between(t_grid, z_smooth, 0, where=in_dir,
                    color="#27ae60", alpha=0.20, zorder=2, label="Expected direction")
    ax.fill_between(t_grid, z_smooth, 0, where=opp_dir,
                    color="#e74c3c", alpha=0.20, zorder=2, label="Opposing direction")

    # ── Phase boundaries + mean lines ────────────────────────────────────
    for ph_name in phase_order:
        interval = phases.get(ph_name)
        if interval is None: continue
        ts, te = interval
        pcol = _PHASE_COLORS[ph_name]
        ax.axvline(ts, color=pcol, lw=1.0, ls="--", alpha=0.45)
        if items:
            ph_vals = [(x[val_key] - mu) / sd if sd > 1e-6 else 0.0
                       for x in items if x.get("phase") == ph_name]
            if len(ph_vals) >= 2:
                ax.hlines(float(np.mean(ph_vals)), ts, te, colors=pcol, lw=2.5,
                          label=f"{_PHASE_LABELS[ph_name]}: {np.mean(ph_vals):+.2f}")

    ax.axhline(0, color="gray", lw=0.7, ls=":")
    ax.set_xlabel("Time in weight session (s)", fontsize=9)
    ax.set_ylabel("Normalised value (z / ±1)", fontsize=9)
    ax.set_ylim(-2.2, 2.2)
    ax.legend(fontsize=7, loc="upper right", ncol=2)
    ax.grid(ls="--", alpha=0.20)
    return r_val, t_grid


def plot_phase_bias(pid, ma_window):
    r = compute_phase_z(pid)
    if r is None:
        print(f"P{pid}: no data"); return None

    phases    = r["phases"]
    direction = r["direction"]
    category, cat_color = _classify_behaviour(r)
    ds_heel   = +1 if direction == "increasing" else -1   # heel: heavier → longer (+1)
    ds_swing  = -ds_heel                                   # swing: heavier → less dynamic (-1)

    # ── Heel contact data ─────────────────────────────────────────────────
    path = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
    df_p, wc = parse_weight_file(path)
    t_s = df_p["time_s"].to_numpy()
    baseline_mask = np.abs(wc) <= _WC_BASE_THR

    heel_items = []
    for col in ("L_heel", "R_heel"):
        contacts, _, _, _ = detect_contacts(df_p[col].to_numpy(), t_s, baseline_mask, wc)
        for c in contacts:
            heel_items.append({"t": c["t_mid"], "dur": c["duration_ms"],
                                "phase": _phase_of(c["t_mid"], phases)})

    # ── Swing acceleration data ───────────────────────────────────────────
    imu_L, imu_R, wc_fn = sync_imu_to_weight_phase(pid)
    sw_items = []
    for imu_df in [imu_L, imu_R]:
        if imu_df is None or imu_df.empty: continue
        if not all(c in imu_df.columns for c in ("ax","ay","az","heel")): continue
        t_i    = imu_df["time_s"].to_numpy()
        heel_i = imu_df["heel"].to_numpy().astype(float)
        net_i  = np.sqrt(imu_df["ax"].to_numpy()**2 +
                         imu_df["ay"].to_numpy()**2 +
                         imu_df["az"].to_numpy()**2).astype(float)
        trigs, rearms = detect_steps_with_dwell(t_i, heel_i)
        for k in range(len(trigs) - 1):
            t0 = t_i[rearms[k]] if k < len(rearms) else t_i[trigs[k]]
            t1 = t_i[trigs[k+1]]
            mk = (t_i >= t0) & (t_i <= t1)
            if mk.sum() < 3: continue
            t_mid = (t0 + t1) / 2
            sw_items.append({"t": t_mid, "acc": float(net_i[mk].max()),
                             "phase": _phase_of(t_mid, phases)})

    # ── Phase bar chart data ──────────────────────────────────────────────
    phase_order = ["baseline_1","ramp_up","plateau","ramp_down","baseline_2"]
    if heel_items:
        all_dur = np.array([x["dur"] for x in heel_items])
        mu_h, sd_h = all_dur.mean(), all_dur.std()
        heel_means = [float(np.mean([(x["dur"]-mu_h)/sd_h if sd_h>1e-6 else 0.0
                                    for x in heel_items if x.get("phase")==p]))
                     if any(x.get("phase")==p for x in heel_items) else 0.0
                     for p in phase_order]
    else:
        heel_means = [0.0] * 5
    if sw_items:
        all_acc = np.array([x["acc"] for x in sw_items])
        mu_s, sd_s = all_acc.mean(), all_acc.std()
        sw_means = [float(np.mean([(x["acc"]-mu_s)/sd_s if sd_s>1e-6 else 0.0
                                   for x in sw_items if x.get("phase")==p]))
                   if any(x.get("phase")==p for x in sw_items) else 0.0
                   for p in phase_order]
    else:
        sw_means = None

    # ── Figure: 3 rows ────────────────────────────────────────────────────
    fig, axes = plt.subplots(3, 1, figsize=(13, 11))
    fig.subplots_adjust(hspace=0.55)

    # Suptitle + category badge
    heel_arrow = "↑" if ds_heel > 0 else "↓"
    sw_arrow   = "↑" if ds_swing > 0 else "↓"
    fig.suptitle(
        f"P{pid}  ·  {direction.title()} weight  "
        f"(Heel expected {heel_arrow},  Swing expected {sw_arrow})",
        fontsize=10, fontweight="bold", y=0.995
    )
    fig.text(0.5, 0.968,
             f"  {category}  ",
             ha="center", va="center", fontsize=10, fontweight="bold",
             color="white",
             bbox=dict(facecolor=cat_color, edgecolor="none",
                       boxstyle="round,pad=0.3", alpha=0.95))

    # Panel 1: Heel contact
    ax1 = axes[0]
    if heel_items:
        r_h, t_grid = _draw_metric_panel(
            ax1, heel_items, "dur", t_s, wc, ds_heel,
            "Heel contact duration", phases)
    else:
        ax1.text(0.5, 0.5, "No heel contact data",
                 ha="center", va="center", transform=ax1.transAxes, color="gray")
        t_grid = None
    ax1.set_title(f"Heel contact duration  ({heel_arrow} expected)", fontsize=10, pad=4)

    # Panel 2: Swing acceleration
    ax2 = axes[1]
    if sw_items:
        _draw_metric_panel(
            ax2, sw_items, "acc", t_s, wc, ds_swing,
            "Swing net acceleration", phases, t_grid=t_grid)
    else:
        ax2.text(0.5, 0.5, "No IMU data available",
                 ha="center", va="center", transform=ax2.transAxes,
                 color="gray", fontsize=11)
    ax2.set_title(f"Swing-phase peak net acceleration  ({sw_arrow} expected,  IMU)",
                  fontsize=10, pad=4)

    # Panel 3: Phase bar chart — peak force (primary) + duration (secondary)
    ax3 = axes[2]
    x  = np.arange(len(phase_order))
    w  = 0.25
    # Peak force bars (primary — significant metric)
    force_means = [r.get("z_" + {"baseline_1":"base1","ramp_up":"ramp_up","plateau":"plateau",
                                  "ramp_down":"ramp_down","baseline_2":"base2"}[p], 0.0)
                   or 0.0 for p in phase_order]
    # Duration bars (secondary)
    dur_means   = [r.get("z_" + {"baseline_1":"base1","ramp_up":"ramp_up","plateau":"plateau",
                                  "ramp_down":"ramp_down","baseline_2":"base2"}[p] + "_dur", 0.0)
                   or 0.0 for p in phase_order]

    for xi, fv in enumerate(force_means):
        ax3.bar(xi - w, fv, width=w, color="#e74c3c", alpha=0.80,
                label="Peak heel force (primary)" if xi == 0 else "", zorder=3)
    for xi, dv in enumerate(dur_means):
        ax3.bar(xi,     dv, width=w, color="#1f77b4", alpha=0.60,
                label="Heel duration (secondary)" if xi == 0 else "", zorder=3)
    if sw_means:
        for xi, sv in enumerate(sw_means):
            ax3.bar(xi + w, sv, width=w, color="#d62728", alpha=0.45,
                    label="Swing accel" if xi == 0 else "", zorder=3)
    ax3.axhspan(0,  2.0, color="#27ae60", alpha=0.04, zorder=0)
    ax3.axhspan(-2.0, 0, color="#e74c3c", alpha=0.04, zorder=0)
    ax3.axhline(0, color="black", lw=0.8)
    ax3.set_xticks(x)
    ax3.set_xticklabels([_PHASE_LABELS[p] for p in phase_order], fontsize=9)
    ax3.set_ylabel("Mean z-score", fontsize=9)
    ax3.set_ylim(-2.0, 2.0)
    ax3.set_title(
        f"Phase summary  |  Force {heel_arrow} expected  "
        f"(red=peak force, blue=duration, pink=swing)",
        fontsize=9, pad=4)
    ax3.legend(fontsize=7, loc="upper right")
    ax3.grid(axis="y", ls="--", alpha=0.3)

    # ── Stats annotation box (Figure 3 equivalent) ───────────────────────
    # Compute influence features for this session
    ds_stats = +1 if direction == "increasing" else -1
    c_early  = r["ramp_response"]   * ds_stats
    c_mid    = r["plateau_response"] * ds_stats
    p_drop   = c_early - c_mid
    z_rd     = r.get("z_ramp_down", float("nan"))
    z_b1     = r.get("z_base1",     float("nan"))
    import math
    opp_late_s = (-(z_rd - z_b1) * ds_stats
                  if not math.isnan(z_rd) and not math.isnan(z_b1)
                  else float("nan"))
    inf_score  = float(np.clip(max(c_early, c_mid, 0.0) / 1.5, 0.0, 1.0))
    opp_str = f"{opp_late_s:+.3f}" if not math.isnan(opp_late_s) else "n/a"
    stats_txt = (
        f"Category:        {category}\n"
        f"Influence score: {inf_score:.3f}\n"
        f"---\n"
        f"correct_early:   {c_early:+.3f}\n"
        f"correct_mid:     {c_mid:+.3f}\n"
        f"plateau_drop:    {p_drop:+.3f}\n"
        f"opp_late:        {opp_str}\n"
        f"---\n"
        f"ramp_response:   {r['ramp_response']:+.3f}\n"
        f"plateau_resp:    {r['plateau_response']:+.3f}\n"
        f"bias_index:      {r['bias_index']:+.3f}"
    )
    ax3.text(
        1.02, 1.0, stats_txt,
        transform=ax3.transAxes, fontsize=7.5,
        verticalalignment="top", fontfamily="monospace",
        bbox=dict(boxstyle="round,pad=0.5", facecolor="white",
                  edgecolor=cat_color, alpha=0.92, lw=1.5),
    )

    plt.tight_layout(rect=(0, 0, 0.88, 0.958))
    return fig

def _refresh_pb(_=None):
    import io as _io
    _pb_out.clear_output(wait=True)
    with _pb_out:
        _fig = plot_phase_bias(_pb_pid_dd.value, _pb_ma_sl.value)
        if _fig is not None:
            _buf = _io.BytesIO()
            _fig.savefig(_buf, dpi=150, bbox_inches='tight', format='png')
            _buf.seek(0)
            display(_fig); plt.close(_fig)
            _btn = widgets.Button(description='Export', button_style='info',
                                   icon='download',
                                   layout=widgets.Layout(width='110px'))
            _lbl = widgets.Label('')
            def _do_save(_, buf=_buf, pid=_pb_pid_dd.value):
                _lbl.value = 'Saving…'
                d = DATA_DIR / 'exports' / 'paper_replication'
                d.mkdir(parents=True, exist_ok=True)
                buf.seek(0)
                (d / f'bias_{pid}.png').write_bytes(buf.read())
                _lbl.value = f'✓  bias_{pid}.png'
            _btn.on_click(_do_save)
            display(widgets.HBox([_btn, _lbl]))

_pb_pid_dd.observe(_refresh_pb, names="value")
_pb_ma_sl.observe(_refresh_pb,  names="value")
_refresh_pb()
display(widgets.HBox([_pb_pid_dd, _pb_ma_sl]), _pb_out)


# ─── Group analysis ───────────────────────────────────────────────────────────

INC_COL = "#e67e22"; DEC_COL = "#9b59b6"

# Figure 1: Ramp vs Plateau scatter
fig_s, ax_s = plt.subplots(figsize=(7, 6))
for direction, color in [("increasing", INC_COL), ("decreasing", DEC_COL)]:
    sub = df_phase[df_phase["direction"] == direction].dropna(
        subset=["ramp_response", "plateau_response"])
    if sub.empty: continue
    ax_s.scatter(sub["plateau_response"], sub["ramp_response"],
                 color=color, s=55, alpha=0.8, zorder=3, label=direction.title())
    for _, row in sub.iterrows():
        ax_s.annotate(row["pid"],
                      (row["plateau_response"], row["ramp_response"]),
                      fontsize=7, alpha=0.7,
                      xytext=(4, 3), textcoords="offset points")
# Diagonal = equal response
lims = [ax_s.get_xlim(), ax_s.get_ylim()]
all_lim = [min(lims[0][0], lims[1][0]) - 0.05, max(lims[0][1], lims[1][1]) + 0.05]
ax_s.plot(all_lim, all_lim, "k--", lw=1.0, alpha=0.5, label="Equal response")
ax_s.set_xlim(all_lim); ax_s.set_ylim(all_lim)
ax_s.axhline(0, color="gray", lw=0.5); ax_s.axvline(0, color="gray", lw=0.5)
ax_s.set_xlabel("Plateau response (z)", fontsize=9)
ax_s.set_ylabel("Ramp response (z)", fontsize=9)
ax_s.set_title("Ramp vs Plateau response\nPoints above diagonal = directional bias",
               fontsize=9)
ax_s.legend(fontsize=8); ax_s.grid(linestyle="--", alpha=0.3)
plt.tight_layout(); plt.show()

# Statistical test: ramp_response vs plateau_response (paired Wilcoxon)
print("\n=== Directional bias statistics ===")
ALPHA_BIAS = 0.05
for direction in ("increasing", "decreasing"):
    sub = df_phase[df_phase["direction"] == direction].dropna(
        subset=["ramp_response", "plateau_response"])
    n = len(sub)
    if n < 3:
        print(f"\n{direction.upper()}: insufficient data (n={n})"); continue
    try:
        _, p = stats.wilcoxon(sub["ramp_response"], sub["plateau_response"],
                               zero_method="zsplit")
    except ValueError:
        print(f"\n{direction.upper()}: Wilcoxon failed"); continue
    z = stats.norm.ppf(p / 2); r = abs(z) / np.sqrt(n)
    sig = "***" if p<0.001 else ("**" if p<0.01 else ("*" if p<ALPHA_BIAS else "ns"))
    bias_med = sub["bias_index"].median()
    print(f"\n{direction.upper()}  (n={n})")
    print(f"  Ramp response:    median={sub['ramp_response'].median():+.3f}  "
          f"IQR=[{sub['ramp_response'].quantile(0.25):+.3f}, {sub['ramp_response'].quantile(0.75):+.3f}]")
    print(f"  Plateau response: median={sub['plateau_response'].median():+.3f}  "
          f"IQR=[{sub['plateau_response'].quantile(0.25):+.3f}, {sub['plateau_response'].quantile(0.75):+.3f}]")
    print(f"  Bias index:       median={bias_med:+.3f}")
    print(f"  Wilcoxon (ramp vs plateau): p={p:.4f} {sig}  r={r:.3f}")
    n_bias = int((sub["bias_index"] > 0).sum())
    print(f"  Directional bias (index > 0): {n_bias}/{n} participants")

# Figure 2: Group trajectory bar chart (B1 → Ramp↑ → Plateau → Ramp↓ → B2)
_phase_cols = ["z_base1", "z_ramp_up", "z_plateau", "z_ramp_down", "z_base2"]
_phase_xlbls = ["B₁\n(pre)", "Ramp↑", "Plateau", "Ramp↓", "B₂\n(post)"]
_phase_plot_colors = ["#aaaaaa", "#e67e22", "#2ecc71", "#3498db", "#aaaaaa"]

fig_t, axes_t = plt.subplots(1, 2, figsize=(12, 5), sharey=True)
fig_t.subplots_adjust(wspace=0.12)
for ax, (direction, color) in zip(axes_t, [("increasing", INC_COL), ("decreasing", DEC_COL)]):
    sub = df_phase[df_phase["direction"] == direction]
    if sub.empty: ax.set_title(f"{direction.title()} (no data)"); continue
    means = [sub[c].mean() for c in _phase_cols]
    sems  = [sub[c].std(ddof=1) / np.sqrt(sub[c].notna().sum()) for c in _phase_cols]
    bars  = ax.bar(range(5), means, yerr=sems, width=0.55,
                   color=_phase_plot_colors, alpha=0.6, capsize=5,
                   error_kw=dict(lw=1.5, capthick=1.5))
    # Individual lines
    for _, row in sub.iterrows():
        vals = [row[c] for c in _phase_cols]
        ax.plot(range(5), vals, "o-", color=color, alpha=0.2, lw=0.8, ms=4)
    ax.axhline(0, color="gray", lw=0.6, linestyle=":")
    ax.set_xticks(range(5)); ax.set_xticklabels(_phase_xlbls, fontsize=8)
    ax.set_ylabel("Mean z-scored peak heel force (primary metric)", fontsize=8)
    ax.set_title(f"{direction.title()} weight  (n={len(sub)})", fontsize=10)
    ax.grid(axis="y", linestyle="--", alpha=0.3)
fig_t.suptitle("Group trajectory across weight session phases\n"
               "Z-scored peak heel force (primary metric) — mean +/- SEM + individual traces",
               fontsize=10)
plt.tight_layout()

# Export
_bias_figs = [(fig_s, "fig_bias_scatter"), (fig_t, "fig_bias_trajectory")]
bias_exp_btn = widgets.Button(description="Export bias plots", button_style="info",
                               icon="download", layout=widgets.Layout(width="155px"))
bias_exp_lbl = widgets.Label("")
def _bias_export(_):
    bias_exp_lbl.value = "Saving…"
    d = DATA_DIR / "exports" / "paper_replication"
    d.mkdir(parents=True, exist_ok=True)
    for fig, fname in _bias_figs:
        fig.savefig(d / f"{fname}.png", dpi=150, bbox_inches="tight")
    bias_exp_lbl.value = f"✓ Saved to exports/paper_replication/"
bias_exp_btn.on_click(_bias_export)
display(widgets.HBox([bias_exp_btn, bias_exp_lbl]))
plt.show()


# ─── Interactive participant identifier — directional bias ────────────────────

_db_out = widgets.Output()

_db_pid_dd = widgets.Dropdown(
    options=sorted(df_phase["pid"].unique()),
    description="Highlight:",
    style={"description_width": "70px"},
    layout=widgets.Layout(width="170px"),
)


def _draw_bias_scatter(highlight_pid=None):
    """Redraw the ramp-vs-plateau scatter, optionally highlighting one participant."""
    fig, ax = plt.subplots(figsize=(5, 4.5))

    for direction, color in [("increasing", INC_COL), ("decreasing", DEC_COL)]:
        sub = df_phase[df_phase["direction"] == direction].dropna(
            subset=["ramp_response", "plateau_response"])
        for _, row in sub.iterrows():
            is_hl = (row["pid"] == highlight_pid)
            ec    = "black" if is_hl else "none"
            ms    = 120    if is_hl else 55
            alpha = 1.0    if is_hl else 0.75
            ax.scatter(row["plateau_response"], row["ramp_response"],
                       color=color, s=ms, alpha=alpha,
                       edgecolors=ec, linewidths=1.5, zorder=4 if is_hl else 3)
            ax.annotate(
                row["pid"],
                (row["plateau_response"], row["ramp_response"]),
                fontsize=7 if not is_hl else 9,
                fontweight="bold" if is_hl else "normal",
                color="black" if is_hl else "#444",
                xytext=(4, 3), textcoords="offset points",
            )

    # Diagonal + axes
    lims = [ax.get_xlim(), ax.get_ylim()]
    lo   = min(lims[0][0], lims[1][0]) - 0.05
    hi   = max(lims[0][1], lims[1][1]) + 0.05
    ax.plot([lo, hi], [lo, hi], "k--", lw=1.0, alpha=0.5, label="Equal response")
    ax.set_xlim(lo, hi); ax.set_ylim(lo, hi)
    ax.axhline(0, color="gray", lw=0.5); ax.axvline(0, color="gray", lw=0.5)

    from matplotlib.patches import Patch
    handles = [
        Patch(color=INC_COL, label="Increasing"),
        Patch(color=DEC_COL, label="Decreasing"),
    ]
    ax.legend(handles=handles, fontsize=8)
    ax.set_xlabel("Plateau response (z)", fontsize=9)
    ax.set_ylabel("Ramp (drift-phase) response (z)", fontsize=9)
    ax.set_title(
        "Drift-phase sensitivity scatter\n"
        "Above diagonal = more affected by ramp onset (drift-phase bias)",
        fontsize=9,
    )
    ax.grid(ls="--", alpha=0.3)

    if highlight_pid:
        row = df_phase[df_phase["pid"] == highlight_pid]
        if not row.empty:
            r = row.iloc[0]
            ax.set_title(
                f"Highlighted: P{highlight_pid}  ({r['direction']})  "
                f"ramp={r['ramp_response']:+.3f}  plat={r['plateau_response']:+.3f}  "
                f"bias={r['bias_index']:+.3f}",
                fontsize=9,
            )

    plt.tight_layout()
    return fig


def _refresh_db(_=None):
    import io as _io
    _db_out.clear_output(wait=True)
    with _db_out:
        _fig = _draw_bias_scatter(_db_pid_dd.value)
        _buf = _io.BytesIO()
        _fig.savefig(_buf, dpi=150, bbox_inches="tight", format="png")
        _buf.seek(0)
        display(_fig); plt.close(_fig)
        _btn = widgets.Button(description="Export", button_style="info",
                               icon="download", layout=widgets.Layout(width="110px"))
        _lbl = widgets.Label("")
        def _save(_, buf=_buf, pid=_db_pid_dd.value):
            _lbl.value = "Saving…"
            d = DATA_DIR / "exports" / "paper_replication"
            d.mkdir(parents=True, exist_ok=True)
            buf.seek(0)
            fname = f"bias_scatter_hl_{pid}.png" if pid else "bias_scatter.png"
            (d / fname).write_bytes(buf.read())
            _lbl.value = f"✓  {fname}"
        _btn.on_click(_save)
        display(widgets.HBox([_btn, _lbl]))


_db_pid_dd.observe(_refresh_db, names="value")
_refresh_db()
display(
    widgets.HTML("<b>Click a participant in the dropdown to highlight them in the scatter "
                 "and jump to their individual bias plot above ↑</b>"),
    widgets.HBox([_db_pid_dd]),
    _db_out,
)
````

## Self-correction analysis — weight & tempo

**Methodology** (consistent across both conditions):

For each participant, the manipulation phase is split into an **early half** and a **late half** by time:

| Condition | Manipulation phase | Early half | Late half |
|---|---|---|---|
| Weight | Plateau (`\|wc\| ≥ 11.4`, ~60 s) | First 50 % of plateau | Last 50 % of plateau |
| Tempo | Active ramp (`\|sf−1\| > 0.01`, ~90 % of session) | First 50 % of ramp | Last 50 % of ramp |

**Self-correction index** = `early_response − late_response`
- **> 0**: gait metric declines during manipulation → self-correction / adaptation
- **≈ 0**: stable response throughout manipulation
- **< 0**: growing response during manipulation (sensitisation)

Metrics applied identically in both conditions:
- Z-scored **heel contact duration** (from pressure insoles)
- Z-scored **swing net acceleration** √(ax²+ay²+az²) (from synced IMU)

> The tempo condition is included as a benchmark: cadence self-correction is established in
> the tempo literature, so if the methodology is sensitive it should detect it
> in the biomechanical metrics here too. The weight condition is expected to show
> a more stable (near-zero self-correction) pattern, given the auditory manipulation
> is subtler and participants are not consciously adjusting walking speed.

````python
# ─── Self-correction helpers ──────────────────────────────────────────────────

_SF_BASE_THR  = 0.01   # |sf - 1.0| <= this  → tempo baseline
_SF_MANIP_THR = 0.01   # |sf - 1.0| >  this  → tempo manipulation active


def parse_tempo_file(path):
    """Parse _pressure_tempo.csv → (df_pressure, sf_array).  Last column = scale_factor."""
    path = Path(path)
    if not path.exists():
        return pd.DataFrame(), np.array([])
    rows_p, sf_list = [], []
    try:
        with open(path, encoding="utf-8", errors="ignore") as fh:
            for line in fh:
                line = line.strip()
                if not line:
                    continue
                tokens = line.rstrip(";").rstrip().split()
                if len(tokens) < 3:
                    continue
                try:
                    sf = float(tokens[-1])
                except ValueError:
                    continue
                unsigned_nums = re.findall(r"[\d]+\.?[\d]*", line)
                if len(unsigned_nums) < 6:
                    continue
                try:
                    ts_ms   = float(unsigned_nums[1])
                    sensors = [float(n) for n in unsigned_nums[1:] if float(n) <= 1023]
                    if len(sensors) < 4:
                        continue
                    rows_p.append({"time_s": ts_ms / 1000.0,
                                   "L_heel": sensors[1], "R_heel": sensors[3]})
                    sf_list.append(sf)
                except (ValueError, IndexError):
                    continue
    except OSError:
        return pd.DataFrame(), np.array([])
    if not rows_p:
        return pd.DataFrame(), np.array([])
    df = pd.DataFrame(rows_p)
    df["time_s"] -= df["time_s"].iloc[0]
    return df, np.array(sf_list)


def sync_imu_to_tempo_phase(pid):
    """Cross-correlate IMU heel signal to tempo pressure heel, return synced IMU window."""
    tf_path = DATA_DIR / pid / f"{pid}_pressure_tempo.csv"
    if not tf_path.exists():
        return None, None, None
    df_t, sf = parse_tempo_file(tf_path)
    if df_t.empty or len(sf) == 0:
        return None, None, None
    t_p        = df_t["time_s"].to_numpy()
    l_heel_p   = df_t["L_heel"].to_numpy().astype(float)
    tempo_dur  = float(t_p[-1])

    df_imu_L = load_imu(pid, "L")
    if df_imu_L.empty or "heel" not in df_imu_L.columns:
        return None, None, None
    t_imu    = df_imu_L["time_s"].to_numpy()
    heel_imu = df_imu_L["heel"].to_numpy().astype(float)
    dt_imu   = float(np.median(np.diff(t_imu)))
    if dt_imu <= 0:
        return None, None, None

    n_rs = max(2, int(tempo_dur / dt_imu))
    if n_rs > len(heel_imu):
        return None, None, None
    t_rs      = np.linspace(0, tempo_dur, n_rs)
    heel_p_rs = np.interp(t_rs, t_p, l_heel_p)

    corr     = _sp_correlate(_norm(heel_imu), _norm(heel_p_rs), mode="valid")
    best_lag = int(np.argmax(corr))
    t_start  = float(t_imu[best_lag])
    t_end    = t_start + tempo_dur

    mask_L  = (t_imu >= t_start) & (t_imu <= t_end)
    df_L_w  = df_imu_L[mask_L].copy().reset_index(drop=True)
    df_L_w["time_s"] -= t_start

    df_imu_R = load_imu(pid, "R")
    df_R_w   = None
    if not df_imu_R.empty and "heel" in df_imu_R.columns:
        t_r   = df_imu_R["time_s"].to_numpy()
        mk_r  = (t_r >= t_start) & (t_r <= t_end)
        df_R_w = df_imu_R[mk_r].copy().reset_index(drop=True)
        df_R_w["time_s"] -= t_start

    def sf_fn(t_local):
        return np.interp(t_local, t_p, sf, left=1.0, right=1.0)

    return df_L_w, df_R_w, sf_fn


def _self_correction_from_contacts(contacts, t_manip_start, t_manip_end, t_base_end):
    """
    Given a list of contacts (each has t_mid, duration_ms) and the time boundaries
    of the baseline and manipulation periods, compute:
      - z-scores within participant (all contacts pooled)
      - mean z in baseline, early manipulation, late manipulation
      - self-correction index = z_early_manip - z_late_manip
    Returns dict or None.
    """
    if len(contacts) < 6:
        return None
    all_dur   = np.array([c["duration_ms"] for c in contacts])
    all_force = np.array([c["peak"]        for c in contacts])
    if all_dur.std() < 1e-6 or all_force.std() < 1e-6:
        return None
    z_all       = (all_dur   - all_dur.mean())   / all_dur.std()
    z_all_force = (all_force - all_force.mean()) / all_force.std()

    t_manip_mid = (t_manip_start + t_manip_end) / 2

    def _seg(zv):
        b = [z for c, z in zip(contacts, zv) if c["t_mid"] <= t_base_end]
        e = [z for c, z in zip(contacts, zv) if t_manip_start <= c["t_mid"] <= t_manip_mid]
        l = [z for c, z in zip(contacts, zv) if t_manip_mid < c["t_mid"] <= t_manip_end]
        return b, e, l

    bf, ef, lf = _seg(z_all_force)
    bd, ed, ld = _seg(z_all)

    if len(ef) < 2 or len(lf) < 2:
        return None

    z_base  = float(np.mean(bf)) if bf else np.nan
    z_early = float(np.mean(ef))
    z_late  = float(np.mean(lf))
    early_resp = z_early - z_base
    late_resp  = z_late  - z_base
    sc_index   = early_resp - late_resp   # primary: peak force

    zbd = float(np.mean(bd)) if bd else np.nan
    sc_index_dur = ((float(np.mean(ed)) - zbd) - (float(np.mean(ld)) - zbd)
                    if len(ed) >= 2 and len(ld) >= 2 else np.nan)

    return {
        "z_baseline": z_base, "z_early": z_early, "z_late": z_late,
        "early_response": early_resp, "late_response": late_resp,
        "sc_index":     sc_index,
        "sc_index_dur": sc_index_dur,
        "contacts": [(c["t_mid"], z) for c, z in zip(contacts, z_all_force)],
        "t_manip_start": t_manip_start, "t_manip_end": t_manip_end,
        "t_manip_mid": t_manip_mid,
    }


def compute_weight_sc(pid):
    """Compute weight self-correction: early vs late plateau for heel contact."""
    direction = get_weight_direction(pid)
    if not direction or direction.lower() == "nan":
        return None
    order = get_condition_order(pid)

    path = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
    df_p, wc = parse_weight_file(path)
    if df_p.empty or len(wc) == 0:
        return None

    t_s           = df_p["time_s"].to_numpy()
    baseline_mask = np.abs(wc) <= _WC_BASE_THR
    phases        = classify_phases(wc, t_s)
    plat          = phases.get("plateau")
    base1         = phases.get("baseline_1")
    if plat is None or base1 is None:
        return None

    # Collect L + R contacts
    all_contacts = []
    for col in ("L_heel", "R_heel"):
        cs, _, _, _ = detect_contacts(df_p[col].to_numpy(), t_s, baseline_mask, wc)
        all_contacts.extend(cs)
    all_contacts.sort(key=lambda c: c["t_mid"])

    res = _self_correction_from_contacts(
        all_contacts,
        t_manip_start=plat[0], t_manip_end=plat[1],
        t_base_end=base1[1]
    )
    if res is None:
        return None
    res.update({"pid": pid, "direction": direction, "order": order, "condition": "weight"})
    return res


def compute_tempo_sc(pid):
    """Compute tempo self-correction: early vs late manipulation phase for heel contact."""
    if pid not in _summary.index:
        return None
    direction_row = str(_summary.loc[pid, "tempo direction"]).strip().replace("_", " ")
    if not direction_row:
        return None
    order = get_condition_order(pid)

    path = DATA_DIR / pid / f"{pid}_pressure_tempo.csv"
    df_t, sf = parse_tempo_file(path)
    if df_t.empty or len(sf) == 0:
        return None

    t_s           = df_t["time_s"].to_numpy()
    baseline_mask = np.abs(sf - 1.0) <= _SF_BASE_THR

    # Find manipulation phase boundaries
    manip_mask = np.abs(sf - 1.0) > _SF_MANIP_THR
    if manip_mask.sum() < 20:
        return None
    t_manip = t_s[manip_mask]
    t_manip_start, t_manip_end = float(t_manip[0]), float(t_manip[-1])
    t_base_end = float(t_s[~manip_mask & (t_s <= t_manip_start)].max()) if (
        (~manip_mask & (t_s <= t_manip_start)).any()) else 0.0

    # Collect L + R contacts
    all_contacts = []
    for col in ("L_heel", "R_heel"):
        if col not in df_t.columns:
            continue
        # Use all-sample baseline mask for threshold computation
        cs, _, _, _ = detect_contacts(df_t[col].to_numpy(), t_s, baseline_mask, sf - 1.0)
        all_contacts.extend(cs)
    all_contacts.sort(key=lambda c: c["t_mid"])

    res = _self_correction_from_contacts(
        all_contacts,
        t_manip_start=t_manip_start, t_manip_end=t_manip_end,
        t_base_end=t_base_end
    )
    if res is None:
        return None
    res.update({"pid": pid, "direction": direction_row, "order": order, "condition": "tempo"})
    return res


def build_sc_stats():
    """Build self-correction DataFrame for all participants, both conditions."""
    rows = []
    for pid in participants:
        if pid in _WEIGHT_EXCLUDE:
            continue
        w = compute_weight_sc(pid)
        if w:
            rows.append({k: v for k, v in w.items()
                         if k not in ("contacts", "t_manip_start", "t_manip_end", "t_manip_mid")})
        t = compute_tempo_sc(pid)
        if t:
            rows.append({k: v for k, v in t.items()
                         if k not in ("contacts", "t_manip_start", "t_manip_end", "t_manip_mid")})
    return pd.DataFrame(rows)

print("Self-correction functions defined.")
print("Building SC stats (tempo IMU sync included — may take ~2 min)…")
df_sc = build_sc_stats()
print(f"SC stats: {len(df_sc)} rows  ({df_sc['condition'].value_counts().to_dict()})")
print()

# Print summary
for cond in ("weight", "tempo"):
    print(f"── {cond.upper()} ──")
    sub = df_sc[df_sc["condition"] == cond]
    for grp in sub["direction"].unique():
        sg = sub[sub["direction"] == grp].dropna(subset=["sc_index"])
        n = len(sg)
        if n == 0: continue
        med = sg["sc_index"].median()
        n_pos = int((sg["sc_index"] > 0).sum())
        try:
            _, p = stats.wilcoxon(sg["early_response"].values, sg["late_response"].values,
                                   zero_method="zsplit")
            z = stats.norm.ppf(p/2); r = abs(z)/np.sqrt(n)
            sig = "***" if p<0.001 else ("**" if p<0.01 else ("*" if p<0.05 else "ns"))
        except Exception:
            p, r, sig = np.nan, np.nan, "—"
        print(f"  {grp:15s}: n={n:2d}  SC_index med={med:+.3f}  "
              f"({n_pos}/{n} positive)  p={p:.4f} {sig}  r={r:.3f}")
    print()
````

## Behaviour categorisation — weight condition

Mirrors the approach in `pressure_behavioral_analysis.ipynb` (Figures 3–5).

**Categories** (4, matching the simplified tempo categories):

| Category | Colour | Definition |
|---|---|---|
| Consistent influence | Green | Metric sustained in expected direction through the plateau (`correct_mid` high, `plateau_drop` low) |
| Impact only from drift phases | Orange | Metric responded during the ramp but faded back to baseline at plateau (`correct_early` high, `plateau_drop` high) |
| Subconscious correction | Blue | Metric initially in expected direction during ramp but then opposes during ramp-down |
| No clear effect | Grey | No systematic expected-direction response |

**Influence score** (analogous to tempo's `compute_influence_score`):  
`max(correct_early, correct_mid, 0) / 1.5`, clipped to [0, 1].  
Reflects the largest detected expected-direction response relative to within-participant noise.

**Per-session trajectory viewer**: use the interactive explorer in the Ramp vs Plateau section above —
it already shows the Figure 3 equivalent (3-panel with trapezoid target + green/red fill + category badge).

The two figures below are the **Figure 4 equivalent** (category grid) and
**Figure 5 equivalent** (influence ranking).

````python
# ─── Feature extraction + influence score ───────────────────────────────────

def build_weight_influence_df():
    """
    Build a per-session feature DataFrame analogous to df_tempo in
    pressure_behavioral_analysis.ipynb.
    """
    rows = []
    for pid in participants:
        if pid in _WEIGHT_EXCLUDE:
            continue
        r = compute_phase_z(pid)
        if r is None:
            continue
        ds = +1 if r["direction"] == "increasing" else -1
        correct_early = r["ramp_response"]   * ds
        correct_mid   = r["plateau_response"] * ds
        plateau_drop  = correct_early - correct_mid
        z_rd = r.get("z_ramp_down", float("nan"))
        z_b1 = r.get("z_base1",    float("nan"))
        import math
        opp_late = (-(z_rd - z_b1) * ds
                    if not math.isnan(z_rd) and not math.isnan(z_b1)
                    else float("nan"))
        # Influence score: max response in EITHER onset (ramp) or sustained
        # (plateau) phase, direction-corrected and normalised to [0,1].
        # Using manipulation-window thirds (not Pearson r) because real gait
        # responses may be delayed, onset-only, or not temporally aligned with
        # the exact wc modulator shape — Pearson r would miss those responses.
        influence = float(np.clip(max(correct_early, correct_mid, 0.0) / 1.5, 0.0, 1.0))
        category, cat_color = _classify_behaviour(r)
        rows.append({
            "pid":          pid,
            "direction":    r["direction"],
            "order":        r["order"],
            "category":     category,
            "cat_color":    cat_color,
            "correct_early": correct_early,
            "correct_mid":  correct_mid,
            "plateau_drop": plateau_drop,
            "opp_late":     opp_late,
            "influence_score": influence,
        })
    return pd.DataFrame(rows)


df_wt_inf = build_weight_influence_df()
print(f"Weight influence table: {len(df_wt_inf)} sessions")
print(df_wt_inf.groupby("category")[["influence_score","correct_early","correct_mid"]]
              .median().round(3).to_string())


# ─── Figure 4 equivalent: category grid ──────────────────────────────────────

def build_weight_influence_df_swing():
    """Parallel to build_weight_influence_df() but using swing phase data."""
    rows = []
    for pid in participants:
        if pid in _WEIGHT_EXCLUDE:
            continue
        r = compute_phase_z_swing(pid)
        if r is None:
            continue
        ds_sw    = -1 if r["direction"] == "increasing" else +1  # FLIPPED
        c_early  = r["ramp_response"]   * ds_sw
        c_mid    = r["plateau_response"] * ds_sw
        p_drop   = c_early - c_mid
        z_rd     = r.get("z_ramp_down", float("nan"))
        z_b1     = r.get("z_base1",     float("nan"))
        import math
        opp_late_s = (-(z_rd - z_b1) * ds_sw
                      if not math.isnan(z_rd) and not math.isnan(z_b1)
                      else float("nan"))
        influence = float(np.clip(max(c_early, c_mid, 0.0) / 1.5, 0.0, 1.0))
        category, cat_color = _classify_behaviour(r)
        rows.append({
            "pid":          pid,
            "direction":    r["direction"],
            "order":        r["order"],
            "category":     category,
            "cat_color":    cat_color,
            "correct_early": c_early,
            "correct_mid":  c_mid,
            "plateau_drop": p_drop,
            "opp_late":     opp_late_s,
            "influence_score": influence,
        })
    return pd.DataFrame(rows)


_CAT_OPTIONS = [cat for cat in _CAT_COLORS if cat in df_wt_inf["category"].values]

_cat_dd  = widgets.Dropdown(
    options=_CAT_OPTIONS if _CAT_OPTIONS else list(_CAT_COLORS.keys()),
    description="Category:",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="290px"),
)
_cat_out = widgets.Output()


def _draw_category_grid(category):
    """Mini-plot grid for all sessions in the selected category."""
    sub = df_wt_inf[df_wt_inf["category"] == category].sort_values(
        "influence_score", ascending=False)
    pids = sub["pid"].tolist()
    if not pids:
        print("No sessions in this category."); return None

    ncols = 4
    nrows = int(np.ceil(len(pids) / ncols))
    fig, axes = plt.subplots(nrows, ncols, figsize=(3.5 * ncols, 3.0 * nrows),
                             squeeze=False)
    for ax in axes.ravel():
        ax.axis("off")

    COMMON = np.linspace(0, 1, 200)
    cat_color = _CAT_COLORS.get(category, "#888")

    for ax, pid in zip(axes.ravel(), pids):
        ax.axis("on")
        row = sub[sub["pid"] == pid].iloc[0]
        direction = row["direction"]
        ds = +1 if direction == "increasing" else -1

        # Load data
        path = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
        df_p, wc = parse_weight_file(path)
        if df_p.empty:
            ax.set_title(f"P{pid}  (no data)", fontsize=8); continue
        t_s = df_p["time_s"].to_numpy()
        baseline_mask = np.abs(wc) <= _WC_BASE_THR

        # Heel contacts
        all_contacts = []
        for col in ("L_heel", "R_heel"):
            cs, _, _, _ = detect_contacts(df_p[col].to_numpy(), t_s, baseline_mask, wc)
            all_contacts.extend(cs)
        if not all_contacts:
            ax.set_title(f"P{pid}  (no contacts)", fontsize=8); continue

        # Z-score + normalise time
        all_contacts.sort(key=lambda c: c["t_mid"])
        t_c   = np.array([c["t_mid"]       for c in all_contacts])
        dur_c   = np.array([c["duration_ms"] for c in all_contacts], dtype=float)
        force_c = np.array([c["peak"]        for c in all_contacts], dtype=float)
        # Use peak force as primary (significant metric) for category grid
        mu, sd = force_c.mean(), force_c.std()
        z_c  = (force_c - mu) / sd if sd > 1e-6 else np.zeros_like(force_c)
        t_n  = t_c / t_s[-1]
        z_on_grid = np.interp(COMMON, t_n, z_c)
        from scipy.ndimage import gaussian_filter1d as _gf
        z_sm = _gf(z_on_grid.astype(float), sigma=4)

        # Target
        wc_norm = np.interp(COMMON, t_s / t_s[-1], (np.abs(wc) / 12.0) * ds)

        # Fill + lines
        in_dir  = (z_sm * ds) > 0
        opp_dir = (z_sm * ds) < 0
        ax.fill_between(COMMON * 100, z_sm, 0, where=in_dir,
                        color="#27ae60", alpha=0.22)
        ax.fill_between(COMMON * 100, z_sm, 0, where=opp_dir,
                        color="#e74c3c", alpha=0.22)
        ax.plot(COMMON * 100, z_sm,   color=cat_color, lw=1.4, alpha=0.9)
        ax.plot(COMMON * 100, wc_norm, color="#2c3e50", lw=1.2,
                ls="--", alpha=0.7)
        ax.axhline(0, color="gray", lw=0.6, ls=":")
        ax.set_ylim(-2.0, 2.0)
        ax.set_xticks([]); ax.set_yticks([])
        ax.set_title(
            "P" + pid + "  " + direction[:3] + "  s=" + f"{row['influence_score']:.2f}",
            fontsize=8, color=cat_color
        )
        ax.grid(ls="--", alpha=0.15)

    fig.suptitle(
        category + " (n=" + str(len(pids)) + ")  "
        "| solid = heel contact z  dashed = wc target  "
        "green = expected dir  red = opposing",
        fontsize=9, y=1.01,
    )
    plt.tight_layout()
    return fig


def _refresh_cat(_=None):
    import io as _io
    _cat_out.clear_output(wait=True)
    with _cat_out:
        _fig = _draw_category_grid(_cat_dd.value)
        if _fig is None: return
        _buf = _io.BytesIO()
        _fig.savefig(_buf, dpi=150, bbox_inches="tight", format="png")
        _buf.seek(0)
        display(_fig); plt.close(_fig)
        _btn = widgets.Button(description="Export", button_style="info",
                               icon="download", layout=widgets.Layout(width="110px"))
        _lbl = widgets.Label("")
        def _save(_, buf=_buf, cat=_cat_dd.value):
            _lbl.value = "Saving..."
            d = DATA_DIR / "exports" / "paper_replication"
            d.mkdir(parents=True, exist_ok=True)
            buf.seek(0)
            fname = "cat_grid_" + cat.replace(" ", "_")[:20] + ".png"
            (d / fname).write_bytes(buf.read())
            _lbl.value = "Saved: " + fname
        _btn.on_click(_save)
        display(widgets.HBox([_btn, _lbl]))


_cat_dd.observe(_refresh_cat, names="value")
_refresh_cat()
display(widgets.HBox([_cat_dd]), _cat_out)


# ─── Figure 5 equivalent: influence ranking ───────────────────────────────────

def _draw_influence_ranking(df_inf=None, title_suffix="Heel contact"):
    """Horizontal bar chart ranking sessions by influence score."""
    if df_inf is None:
        df_inf = df_wt_inf
    df_sorted = df_inf.sort_values("influence_score", ascending=True)
    n = len(df_sorted)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, max(4, n * 0.38 + 1)),
                                    gridspec_kw={"width_ratios": [3, 1]})
    fig.subplots_adjust(wspace=0.35)
    y_pos = np.arange(n)
    for i, (_, row) in enumerate(df_sorted.iterrows()):
        ax1.barh(i, row["influence_score"], color=row["cat_color"], alpha=0.80, height=0.7)
        lbl = "P" + row["pid"] + "  " + row["category"][:14]
        ax1.text(row["influence_score"] + 0.01, i, lbl, va="center", fontsize=7.5)
    for x, label in [(0.15, "Weak"), (0.40, "Moderate")]:
        ax1.axvline(x, color="#aaa", lw=1.0, ls="--", alpha=0.7)
        ax1.text(x + 0.005, n - 0.3, label, fontsize=7, color="#888")
    ax1.text(0.42, n - 0.3, "Strong", fontsize=7, color="#888")
    ax1.set_yticks(y_pos)
    ax1.set_yticklabels(df_sorted["direction"].str[:3].values, fontsize=8)
    ax1.set_xlabel("Influence score  (0=none, 1=strong)", fontsize=9)
    ax1.set_xlim(0, 1.15)
    ax1.set_title("Sessions ranked by influence score\n(" + title_suffix + ")", fontsize=10)
    ax1.grid(axis="x", ls="--", alpha=0.3)
    cat_summary = []
    for cat in list(_CAT_COLORS.keys()):
        sub = df_inf[df_inf["category"] == cat]
        if sub.empty: continue
        cat_summary.append({
            "Category": cat[:22], "_cat_full": cat,
            "n": len(sub),
            "Med score": round(float(sub["influence_score"].median()), 2),
            "Med early": round(float(sub["correct_early"].median()), 2),
            "Med mid":   round(float(sub["correct_mid"].median()),   2),
        })
    tdf = pd.DataFrame(cat_summary)
    ax2.axis("off")
    if not tdf.empty:
        tbl = ax2.table(cellText=tdf[["Category","n","Med score","Med early","Med mid"]].values,
                        colLabels=["Category","n","Med score","Med early","Med mid"],
                        cellLoc="center", loc="center")
        tbl.auto_set_font_size(False); tbl.set_fontsize(8); tbl.scale(1, 1.4)
        for i, row in enumerate(cat_summary):
            col = _CAT_COLORS.get(row["_cat_full"], "#eeeeee")
            for j in range(5):
                tbl[i+1, j].set_facecolor(col + "44")
    ax2.set_title("Category summary", fontsize=9, pad=12)
    plt.tight_layout()
    return fig


def _draw_influence_ranking_both():
    """Side-by-side influence ranking: heel contact (left) vs swing accel (right)."""
    fig, axes = plt.subplots(1, 2, figsize=(16, max(5, len(df_wt_inf) * 0.42 + 1)))
    fig.subplots_adjust(wspace=0.55)

    def _half(ax, df_inf, title):
        df_sorted = df_inf.sort_values("influence_score", ascending=True)
        n = len(df_sorted)
        for i, (_, row) in enumerate(df_sorted.iterrows()):
            ax.barh(i, row["influence_score"], color=row["cat_color"], alpha=0.80, height=0.7)
            ax.text(row["influence_score"] + 0.01, i,
                    "P" + row["pid"] + "  " + row["category"][:12],
                    va="center", fontsize=7)
        for x, lbl in [(0.15, "Weak"), (0.40, "Mod")]:
            ax.axvline(x, color="#aaa", lw=1.0, ls="--", alpha=0.7)
            ax.text(x + 0.005, n - 0.3, lbl, fontsize=7, color="#888")
        ax.set_yticks(range(n))
        ax.set_yticklabels(df_sorted["direction"].str[:3].values, fontsize=8)
        ax.set_xlabel("Influence score", fontsize=9)
        ax.set_xlim(0, 1.15)
        ax.set_title(title, fontsize=10)
        ax.grid(axis="x", ls="--", alpha=0.3)

    _half(axes[0], df_wt_inf,    "Heel contact duration")
    if len(df_wt_inf_sw) > 0:
        _half(axes[1], df_wt_inf_sw, "Swing-phase acceleration")
    else:
        axes[1].text(0.5, 0.5, "No swing data", ha="center", va="center",
                     transform=axes[1].transAxes, color="gray", fontsize=12)
    fig.suptitle("Behaviour influence ranking — heel vs swing comparison", fontsize=11)
    plt.tight_layout()
    return fig






print('Building swing behaviour stats (IMU sync ~2 min)...')
df_wt_inf_sw = build_weight_influence_df_swing()
print(f'  Swing sessions: {len(df_wt_inf_sw)}')

_fig_inf      = _draw_influence_ranking(df_wt_inf,    title_suffix='Heel contact')
_fig_inf_sw   = _draw_influence_ranking(df_wt_inf_sw, title_suffix='Swing acceleration')
_fig_inf_both = _draw_influence_ranking_both()
for fig in [_fig_inf, _fig_inf_sw, _fig_inf_both]:
    plt.figure(fig.number); plt.show()

# Export
import io as _io_inf
_inf_buf = _io_inf.BytesIO()
_fig_inf.savefig(_inf_buf, dpi=150, bbox_inches="tight", format="png")
_inf_buf.seek(0)
_inf_btn = widgets.Button(description="Export ranking", button_style="info",
                           icon="download", layout=widgets.Layout(width="150px"))
_inf_lbl = widgets.Label("")
def _inf_save(_):
    _inf_lbl.value = "Saving..."
    d = DATA_DIR / "exports" / "paper_replication"
    d.mkdir(parents=True, exist_ok=True)
    _inf_buf.seek(0)
    (d / "fig5_weight_influence_ranking.png").write_bytes(_inf_buf.read())
    _inf_lbl.value = "Saved: fig5_weight_influence_ranking.png"
_inf_btn.on_click(_inf_save)
display(widgets.HBox([_inf_btn, _inf_lbl]))


# ─── Heel vs Swing category comparison table ─────────────────────────────────

def _build_comparison_table():
    """
    Merge heel and swing influence DataFrames, one row per session,
    showing how each metric labelled the session.
    """
    if df_wt_inf.empty or df_wt_inf_sw.empty:
        print("Run both df_wt_inf and df_wt_inf_sw first."); return None

    h = df_wt_inf[["pid","direction","order","category","influence_score",
                    "correct_early","correct_mid"]].copy()
    h.columns = ["pid","direction","order",
                  "heel_category","heel_score","heel_early","heel_mid"]

    s = df_wt_inf_sw[["pid","category","influence_score",
                       "correct_early","correct_mid"]].copy()
    s.columns = ["pid","swing_category","swing_score","swing_early","swing_mid"]

    merged = h.merge(s, on="pid", how="outer")
    merged["agree"] = merged["heel_category"] == merged["swing_category"]

    # Sort: heel category alphabetically, then heel_score descending
    cat_order = {"Consistent influence": 0,
                 "Impact only from drift phases": 1,
                 "Subconscious correction": 2,
                 "No clear effect": 3}
    merged["_sort"] = merged["heel_category"].map(cat_order).fillna(4)
    merged = merged.sort_values(["_sort","heel_score"], ascending=[True,False])
    merged = merged.drop(columns="_sort")

    n_agree = int(merged["agree"].sum())
    n_total = len(merged)
    print(f"Agreement: {n_agree}/{n_total} sessions ({100*n_agree/n_total:.0f}%) "
          "categorised identically by both metrics")
    print()

    # Category counts per metric
    for metric, col in [("Heel", "heel_category"), ("Swing", "swing_category")]:
        counts = merged[col].value_counts()
        print(f"{metric}: " + "  |  ".join(
            f"{cat[:20]}: {n}" for cat, n in counts.items()))
    print()

    return merged


def _style_comparison_table(df):
    """Apply background colours and agree/disagree symbols."""
    if df is None:
        return

    def _bg_heel(val):
        col = _CAT_COLORS.get(val, "#eeeeee")
        return f"background-color: {col}44; font-weight: bold"

    def _bg_swing(val):
        col = _CAT_COLORS.get(val, "#eeeeee")
        return f"background-color: {col}44; font-weight: bold"

    def _bg_agree(val):
        if val is True:
            return "background-color: #d5f5e344; color: #1e8449; font-weight: bold"
        elif val is False:
            return "background-color: #fadbd844; color: #c0392b; font-weight: bold"
        return ""

    display_df = df[["pid","direction","heel_category","heel_score",
                      "swing_category","swing_score","agree"]].copy()
    display_df["heel_score"]  = display_df["heel_score"].round(3)
    display_df["swing_score"] = display_df["swing_score"].round(3)
    display_df["agree"]       = display_df["agree"].map({True: "✓", False: "✗"})
    display_df.columns = ["PID","Direction",
                           "Heel category","Heel score",
                           "Swing category","Swing score","Agree?"]

    styled = (display_df.style
              .map(_bg_heel,  subset=["Heel category"])
              .map(_bg_swing, subset=["Swing category"])
              .map(_bg_agree, subset=["Agree?"])
              .set_table_styles([
                  {"selector": "th",
                   "props": [("background-color","#2c3e50"),
                              ("color","white"),("font-weight","bold"),
                              ("padding","6px 10px")]},
                  {"selector": "td",
                   "props": [("padding","5px 10px"),("font-size","11px")]},
              ])
              .hide(axis="index"))
    display(styled)


def _export_comparison_table(df):
    """Save comparison table as a matplotlib figure."""
    if df is None: return

    display_df = df[["pid","direction","heel_category","heel_score",
                      "swing_category","swing_score","agree"]].copy()
    display_df["heel_score"]  = display_df["heel_score"].round(3)
    display_df["swing_score"] = display_df["swing_score"].round(3)
    display_df["agree"]       = display_df["agree"].map({True: "Y", False: "N"})
    display_df.columns = ["PID","Dir",
                           "Heel category","H.score",
                           "Swing category","S.score","=?"]
    n = len(display_df)

    fig, ax = plt.subplots(figsize=(14, max(2.5, n * 0.38 + 0.8)))
    ax.axis("off")
    tbl = ax.table(cellText=display_df.values,
                   colLabels=display_df.columns,
                   cellLoc="center", loc="center")
    tbl.auto_set_font_size(False); tbl.set_fontsize(8.5)
    tbl.scale(1, 1.5)

    # Colour header
    for j in range(len(display_df.columns)):
        tbl[0, j].set_facecolor("#2c3e50")
        tbl[0, j].set_text_props(color="white", fontweight="bold")

    # Colour category cells + agree column
    for i, (_, row) in enumerate(df.iterrows(), start=1):
        h_col = _CAT_COLORS.get(row["heel_category"],  "#eeeeee") + "66"
        s_col = _CAT_COLORS.get(row["swing_category"], "#eeeeee") + "66"
        tbl[i, 2].set_facecolor(h_col)
        tbl[i, 4].set_facecolor(s_col)
        tbl[i, 6].set_facecolor("#d5f5e3" if row["agree"] else "#fadbd8")
        tbl[i, 6].set_text_props(
            color=("#1e8449" if row["agree"] else "#c0392b"), fontweight="bold")

    n_agree = int(df["agree"].sum())
    fig.suptitle(
        "Heel contact vs Swing acceleration — per-session category comparison\n"
        "Agreement: " + str(n_agree) + "/" + str(n) +
        " sessions (" + str(round(100*n_agree/n)) + "%) assigned same category",
        fontsize=10, y=1.01,
    )
    plt.tight_layout()
    return fig


_df_cmp = _build_comparison_table()
_style_comparison_table(_df_cmp)

import io as _io_cmp
_cmp_fig = _export_comparison_table(_df_cmp)
if _cmp_fig:
    plt.show()
    _cmp_buf = _io_cmp.BytesIO()
    _cmp_fig.savefig(_cmp_buf, dpi=150, bbox_inches="tight", format="png")
    _cmp_buf.seek(0)

    _cmp_btn = widgets.Button(description="Export table", button_style="info",
                               icon="download", layout=widgets.Layout(width="130px"))
    _cmp_lbl = widgets.Label("")
    def _cmp_save(_):
        _cmp_lbl.value = "Saving..."
        d = DATA_DIR / "exports" / "paper_replication"
        d.mkdir(parents=True, exist_ok=True)
        _cmp_buf.seek(0)
        (d / "fig_category_comparison_table.png").write_bytes(_cmp_buf.read())
        _cmp_lbl.value = "Saved: fig_category_comparison_table.png"
    _cmp_btn.on_click(_cmp_save)
    display(widgets.HBox([_cmp_btn, _cmp_lbl]))


# ─── Inter-metric correlation: do heel and swing detect the same sessions? ────

def _draw_metric_correlation():
    """
    Scatter: heel influence_r (x) vs swing influence_r (y), one point per session.
    Tests whether participants who showed a heel response also showed a swing response.
    A positive correlation = consistent detectors; near-zero = independent measures.
    """
    if df_wt_inf.empty or df_wt_inf_sw.empty:
        print("Run both df_wt_inf and df_wt_inf_sw first."); return None

    # Merge on pid
    h = df_wt_inf[["pid","direction","influence_score"]].copy()
    h.columns = ["pid","direction","heel_r"]
    s = df_wt_inf_sw[["pid","influence_score"]].copy()
    s.columns = ["pid","swing_r"]
    merged = h.merge(s, on="pid").dropna(subset=["heel_r","swing_r"])

    if len(merged) < 5:
        print("Insufficient data for correlation."); return None

    heel_r  = merged["heel_r"].values
    swing_r = merged["swing_r"].values
    r_val, p_val = stats.pearsonr(heel_r, swing_r)

    fig, ax = plt.subplots(figsize=(6, 5))
    for direction, color in [("increasing","#e67e22"),("decreasing","#9b59b6")]:
        sub = merged[merged["direction"] == direction]
        ax.scatter(sub["heel_r"], sub["swing_r"],
                   color=color, s=55, alpha=0.8, label=direction, zorder=3)
        for _, row in sub.iterrows():
            ax.annotate("P" + row["pid"],
                        (row["heel_r"], row["swing_r"]),
                        fontsize=7, xytext=(3,2), textcoords="offset points",
                        color="#555")

    # Reference lines
    ax.axhline(0, color="gray", lw=0.6, ls=":")
    ax.axvline(0, color="gray", lw=0.6, ls=":")
    lo = min(heel_r.min(), swing_r.min()) - 0.05
    hi = max(heel_r.max(), swing_r.max()) + 0.05
    ax.plot([lo, hi], [lo, hi], "k--", lw=0.8, alpha=0.4, label="x = y (perfect agreement)")

    sig = "***" if p_val<0.001 else ("**" if p_val<0.01 else ("*" if p_val<0.05 else "ns"))
    ax.text(0.04, 0.96,
            "Pearson r = " + f"{r_val:+.3f}" + "  p = " + f"{p_val:.3f}" + " " + sig,
            transform=ax.transAxes, fontsize=9, va="top",
            bbox=dict(facecolor="white", edgecolor="#bbb", alpha=0.9, pad=4))

    ax.set_xlabel("Heel contact influence score", fontsize=9)
    ax.set_ylabel("Swing acceleration influence score", fontsize=9)
    ax.set_title(
        "Inter-metric agreement: heel vs swing\n"
        "r > 0.3 = consistent detectors  |  r near 0 = independent measures",
        fontsize=9)
    ax.legend(fontsize=8); ax.grid(ls="--", alpha=0.3)
    plt.tight_layout()

    # Interpretation
    print("Inter-metric correlation: r = " + f"{r_val:+.3f}" + ", p = " + f"{p_val:.4f}" + " " + sig)
    if abs(r_val) < 0.2:
        print("  Interpretation: Near-zero correlation — heel contact and swing acceleration")
        print("  are detecting INDEPENDENT variance. They measure different aspects of gait")
        print("  (stance loading vs swing propulsion). Use heel as primary (stronger in")
        print("  Tajadura-Jimenez 2015); swing as independent secondary check.")
    elif r_val > 0.3:
        print("  Interpretation: Positive correlation — participants who responded with heel")
        print("  also responded with swing. Both metrics are consistent detectors.")
        print("  Consider averaging the two influence scores for a combined measure.")
    else:
        print("  Interpretation: Weak positive correlation — partial agreement between metrics.")

    return fig


_fig_corr = _draw_metric_correlation()
if _fig_corr:
    import io as _io_c
    plt.show()
    _c_buf = _io_c.BytesIO()
    _fig_corr.savefig(_c_buf, dpi=150, bbox_inches="tight", format="png")
    _c_buf.seek(0)
    _c_btn = widgets.Button(description="Export", button_style="info",
                             icon="download", layout=widgets.Layout(width="110px"))
    _c_lbl = widgets.Label("")
    def _c_save(_):
        _c_lbl.value = "Saving..."
        d = DATA_DIR / "exports" / "paper_replication"
        d.mkdir(parents=True, exist_ok=True)
        _c_buf.seek(0)
        (d / "fig_heel_swing_correlation.png").write_bytes(_c_buf.read())
        _c_lbl.value = "Saved: fig_heel_swing_correlation.png"
    _c_btn.on_click(_c_save)
    display(widgets.HBox([_c_btn, _c_lbl]))
````

## Within-participant z-scoring

Following Tajadura-Jiménez et al. (2015):
> *"Individual z-scores were calculated to reduce intra-subject variability."*

Each participant's per-step measurements are z-scored across **all their steps in the weight session** (pooling baseline and plateau), then averaged within each segment. The result is a per-participant mean z-score for each condition — directly comparable to the paper's Figure 4.

````python
PRESSURE_METRICS = ["heel_contact_ms", "peak_heel_force", "gait_cycle_ms"]
IMU_METRICS      = ["net_accel_peak", "net_accel_rms", "swing_duration_ms"]

def zscore_within(df, metrics):
    df = df.copy()
    for m in metrics:
        if m not in df.columns: continue
        df[f"{m}_z"] = np.nan
        for pid in df["pid"].unique():
            mk = df["pid"] == pid
            vals = df.loc[mk, m].to_numpy(dtype=float)
            mu, sd = float(np.nanmean(vals)), float(np.nanstd(vals))
            if sd > 1e-6:
                df.loc[mk, f"{m}_z"] = (vals - mu) / sd
    return df

def participant_summary(df, metrics):
    rows = []
    for pid in df["pid"].unique():
        sub = df[df["pid"] == pid]
        row = {"pid": pid, "direction": sub["direction"].iloc[0], "order": sub["order"].iloc[0]}
        for m in metrics:
            col = f"{m}_z"
            if col not in sub.columns: continue
            for seg in ("baseline", "plateau"):
                v = sub.loc[sub["wc_segment"] == seg, col].dropna().values
                row[f"{m}_{seg}"] = float(np.mean(v)) if len(v) >= 1 else np.nan
        rows.append(row)
    return pd.DataFrame(rows)

df_steps  = zscore_within(df_steps,  PRESSURE_METRICS)
df_swings = zscore_within(df_swings, IMU_METRICS)

df_psum_p = participant_summary(df_steps,  PRESSURE_METRICS)
df_psum_i = participant_summary(df_swings, IMU_METRICS)

print("Participant-level z-score summary (heel contact):")
print(df_psum_p[["pid","direction","heel_contact_ms_baseline","heel_contact_ms_plateau"]
                ].round(3).to_string(index=False))
````

## Statistical analysis

Following the same approach as the tempo final analysis (`pressure_timeseries.ipynb`):

1. **Wilcoxon signed-rank test** (always non-parametric, matching tempo analysis)
2. **Effect size r** = |Z| / √N — small ≥ 0.1, medium ≥ 0.3, large ≥ 0.5
3. α = **0.05** (no Bonferroni correction, matching tempo)

**Directional hypotheses** (mirroring paper):
- *Increasing* (heavier sound): heel contact time **longer** at plateau
- *Decreasing* (lighter sound): heel contact time **shorter** at plateau; net acceleration **larger** at plateau

````python
ALPHA = 0.05   # matching tempo analysis (no Bonferroni correction)

def run_test(df_psum, metric, direction):
    """Always Wilcoxon signed-rank + effect size r  (mirrors tempo final analysis)."""
    sub = df_psum[df_psum["direction"] == direction].dropna(
        subset=[f"{metric}_baseline", f"{metric}_plateau"])
    n = len(sub)
    if n < 3:
        return {"metric": metric, "direction": direction, "n": n, "sig": "—",
                "p": np.nan, "effect": np.nan, "effect_label": "r",
                "med_base": np.nan, "med_plat": np.nan, "med_delta": np.nan,
                "test": "too few"}
    base  = sub[f"{metric}_baseline"].values
    plat  = sub[f"{metric}_plateau"].values
    delta = plat - base
    try:
        _, p_val = stats.wilcoxon(base, plat, zero_method="zsplit")
    except ValueError:
        return {"metric": metric, "direction": direction, "n": n, "sig": "—",
                "p": np.nan, "effect": np.nan, "effect_label": "r",
                "med_base": float(np.median(base)), "med_plat": float(np.median(plat)),
                "med_delta": float(np.median(delta)), "test": "Wilcoxon failed"}
    z = stats.norm.ppf(p_val / 2)
    r = abs(z) / np.sqrt(n)
    sig = "***" if p_val<0.001 else ("**" if p_val<0.01 else ("*" if p_val<0.05 else "ns"))
    return {"metric": metric, "direction": direction, "n": n,
            "med_base": float(np.median(base)), "med_plat": float(np.median(plat)),
            "med_delta": float(np.median(delta)), "test": "Wilcoxon",
            "p": float(p_val), "effect": float(r), "effect_label": "r", "sig": sig}

all_res = []
print(f"=== Statistical results  (Wilcoxon signed-rank, α={ALPHA}) ===\n")
for df_psum, metrics, label in [(df_psum_p, PRESSURE_METRICS, "Pressure"),
                                  (df_psum_i, IMU_METRICS,      "IMU")]:
    if df_psum.empty: continue
    print(f"── {label} ──")
    for m in metrics:
        for d in ("increasing", "decreasing"):
            r = run_test(df_psum, m, d)
            all_res.append(r)
            print(f"  {m:25s} | {d:11s} | n={r['n']:2d} | "
                  f"Δmed={r['med_delta']:+.3f} | p={r['p']:.4f} {r['sig']:3s} | "
                  f"r={r['effect']:.3f}")
    print()

df_results = pd.DataFrame(all_res)
````

## Appendix: Behaviour categorisation — false positive validation

Three independent null checks confirm that the behaviour categorisation method
is detecting weight-manipulation-specific responses, not general gait variability.

**Expected outcome of all three**: category rates indistinguishable from chance
(~25% per category for a 4-category classifier); unambiguous categories
(Consistent influence + Impact from drift phases + Subconscious correction)
well below the rate observed in weight sessions.

| Method | Dataset | n | Logic |
|---|---|---|---|
| A | T6 tempo sessions (same participants) | 21 | Tempo changes cadence, not heel force → expect null |
| B | T6 weight, permuted timing (×500) | 20 | Same data, manipulation timing removed → true null |
| C | T5 mock-trial sessions (independent round) | 6 | Different participants, earlier round → expect null |

````python
# ─── Generalised phase-z for any session file ────────────────────────────────

def compute_phase_z_generic(path, modulator_fn, direction_sign,
                             neutral_thr=0.01, min_manip=0.01):
    """
    Apply the manipulation-window thirds classification to any session file.
    modulator_fn: callable that converts raw last-column value to a signed
                  'deviation from neutral' (e.g. lambda v: v for wc,
                  or lambda v: v-1.0 for scale_factor).
    direction_sign: +1 or -1 (expected force direction).
    Returns the same dict structure as compute_phase_z().
    """
    df_p, raw_mod = parse_weight_file(path)   # reuses existing parser
    if df_p.empty or len(raw_mod) == 0:
        return None

    mod        = np.array([modulator_fn(v) for v in raw_mod])
    t_s        = df_p["time_s"].to_numpy()
    base_mask  = np.abs(mod) <= neutral_thr
    phases_wc  = classify_phases(mod, t_s)   # still uses mod for vis

    all_contacts = []
    for col in ("L_heel", "R_heel"):
        cs, _, _, _ = detect_contacts(df_p[col].to_numpy(), t_s, base_mask, mod)
        all_contacts.extend(cs)

    if len(all_contacts) < 10:
        return None

    all_force = np.array([c["peak"] for c in all_contacts])
    if all_force.std() < 1e-6:
        return None
    z_f = (all_force - all_force.mean()) / all_force.std()

    import math
    manip_mask = np.abs(mod) > min_manip
    manip_t    = t_s[manip_mask]
    if len(manip_t) < 10:
        return None

    t_ms = float(manip_t[0]); t_me = float(manip_t[-1])
    md_  = t_me - t_ms
    t_e1 = t_ms + md_ / 3.0; t_e2 = t_ms + 2.0 * md_ / 3.0

    thirds = {"baseline": [], "early": [], "mid": [], "late": []}
    for c, z in zip(all_contacts, z_f):
        tm = c["t_mid"]
        if tm < t_ms:       thirds["baseline"].append(z)
        elif tm <= t_e1:    thirds["early"].append(z)
        elif tm <= t_e2:    thirds["mid"].append(z)
        elif tm <= t_me:    thirds["late"].append(z)

    def m(lst): return float(np.mean(lst)) if len(lst) >= 2 else np.nan

    ref   = m(thirds["baseline"]) if len(thirds["baseline"]) >= 3 else 0.0
    ramp  = (m(thirds["early"]) - ref)
    plat  = (m(thirds["mid"])   - ref)
    if math.isnan(ramp) or math.isnan(plat):
        return None
    bias  = ramp - plat

    fake_r = {"direction": ("increasing" if direction_sign > 0 else "decreasing"),
              "ramp_response": ramp * direction_sign,
              "plateau_response": plat * direction_sign,
              "bias_index": bias}
    category, _ = _classify_behaviour(fake_r)
    return category


def _category_rates(cats):
    from collections import Counter
    c = Counter(cats)
    n = len(cats)
    order = ["Consistent influence", "Impact only from drift phases",
             "Subconscious correction", "No clear effect"]
    return {k: (c.get(k, 0), 100*c.get(k,0)/n if n else 0) for k in order}


def _print_rates(label, rates, n):
    print(f"  {label} (n={n})")
    for cat, (cnt, pct) in rates.items():
        print(f"    {cat[:30]:30s}: {cnt:2d} ({pct:.0f}%)")


# ── Method A: T6 tempo sessions (same participants) ───────────────────────────

cats_A = []
for pid in participants:
    if pid in _WEIGHT_EXCLUDE: continue
    path = DATA_DIR / pid / f"{pid}_pressure_tempo.csv"
    if not path.exists(): continue
    # Tempo modulator: scale_factor, deviation = sf-1.0
    # Direction: speeding_up → expected heel force decrease (lighter feel?)
    # We use direction_sign=+1 for all (worst case — maximises false positives)
    cat = compute_phase_z_generic(path,
                                   modulator_fn=lambda v: v - 1.0,
                                   direction_sign=+1,
                                   neutral_thr=0.005, min_manip=0.005)
    if cat:
        cats_A.append(cat)


# ── Method B: Permutation test (T6 weight, shuffle thirds) ───────────────────

import random as _rnd
_rnd.seed(42)
N_PERM    = 500
perm_cats = []

for pid in participants:
    if pid in _WEIGHT_EXCLUDE: continue
    path = DATA_DIR / pid / f"{pid}_pressure_weight.csv"
    df_p, wc = parse_weight_file(path)
    if df_p.empty: continue
    t_s = df_p["time_s"].to_numpy()
    base_mask = np.abs(wc) <= _WC_BASE_THR
    all_contacts = []
    for col in ("L_heel", "R_heel"):
        cs, _, _, _ = detect_contacts(df_p[col].to_numpy(), t_s, base_mask, wc)
        all_contacts.extend(cs)
    if len(all_contacts) < 10: continue
    all_force = np.array([c["peak"] for c in all_contacts])
    if all_force.std() < 1e-6: continue
    z_f = (all_force - all_force.mean()) / all_force.std()
    t_c = np.array([c["t_mid"] for c in all_contacts])
    dur = float(t_s[-1])
    direction = get_weight_direction(pid)
    ds = +1 if direction == "increasing" else -1

    for _ in range(N_PERM // len([p for p in participants if p not in _WEIGHT_EXCLUDE and
                                   (DATA_DIR/p/f"{p}_pressure_weight.csv").exists()])):
        # Random manipulation window same duration as real one
        import math as _m
        manip_dur = dur * 0.58   # ~73s / 125s average
        t_start   = _rnd.uniform(0, dur - manip_dur)
        t_end     = t_start + manip_dur
        t_e1 = t_start + manip_dur / 3.0; t_e2 = t_start + 2.0 * manip_dur / 3.0
        thirds = {"baseline": [], "early": [], "mid": [], "late": []}
        for tc, z in zip(t_c, z_f):
            if tc < t_start:       thirds["baseline"].append(z)
            elif tc <= t_e1:       thirds["early"].append(z)
            elif tc <= t_e2:       thirds["mid"].append(z)
            elif tc <= t_end:      thirds["late"].append(z)
        ref  = float(np.mean(thirds["baseline"])) if len(thirds["baseline"]) >= 3 else 0.0
        ramp = (float(np.mean(thirds["early"])) - ref) if len(thirds["early"]) >= 2 else _m.nan
        plat = (float(np.mean(thirds["mid"]))   - ref) if len(thirds["mid"])   >= 2 else _m.nan
        if _m.isnan(ramp) or _m.isnan(plat): continue
        fake_r = {"direction": direction,
                  "ramp_response":    ramp * ds,
                  "plateau_response": plat * ds,
                  "bias_index":       ramp - plat}
        cat, _ = _classify_behaviour(fake_r)
        perm_cats.append(cat)


# ── Method C: T5 mock-trial (independent round, scale_factor modulator) ──────

T5_DIR = Path("/Users/zoe/Library/CloudStorage/OneDrive-Personal/University/year4/Masters/Final Submission Data analysis/t5_mocktrial")
cats_C = []
for p_dir in sorted(T5_DIR.iterdir()):
    if not p_dir.is_dir(): continue
    # Find any weight or tempo file (both use scale_factor in T5)
    files = list(p_dir.glob("*weight*.csv")) + list(p_dir.glob("*_pressure_weight.csv"))
    for wf in files:
        cat = compute_phase_z_generic(wf,
                                       modulator_fn=lambda v: v - 1.0,
                                       direction_sign=+1,
                                       neutral_thr=0.005, min_manip=0.005)
        if cat:
            cats_C.append(cat)
        break   # one per participant


# ── Comparison ────────────────────────────────────────────────────────────────

# Real weight session rates
cats_real = list(df_wt_inf["category"].values) if "df_wt_inf" in dir() else []

print("=" * 60)
print("BEHAVIOUR CATEGORISATION — FALSE POSITIVE VALIDATION")
print("=" * 60)

if cats_real:
    _print_rates("Weight sessions (REAL)", _category_rates(cats_real), len(cats_real))
    print()

_print_rates("Method A: T6 tempo sessions", _category_rates(cats_A), len(cats_A))
print()
_print_rates("Method B: Permuted timing (x500)", _category_rates(perm_cats), len(perm_cats))
print()
_print_rates("Method C: T5 independent round", _category_rates(cats_C), len(cats_C))

# Unambiguous category rate (the meaningful comparison)
def _unamb_rate(cats):
    n = len(cats)
    if n == 0: return 0, 0
    unamb = sum(1 for c in cats if c != "No clear effect")
    return unamb, 100*unamb/n

print()
print("Unambiguous category rate (Consistent + Drift + Correction):")
for label, cats in [("Real weight", cats_real), ("A: T6 tempo", cats_A),
                     ("B: Permuted", perm_cats), ("C: T5 independent", cats_C)]:
    u, pct = _unamb_rate(cats)
    print(f"  {label:25s}: {u}/{len(cats)} ({pct:.0f}%)")
````
