# T6 — Agency & Influence Correlation Analysis

Examines whether participants' sense of agency relates to how much the weight manipulation influenced their gait. The primary question is:

> **Do participants who felt more in control of their walking show stronger (or weaker) gait changes in response to the weight sound manipulation?**

## Variables
| Variable | Description |
|---|---|
| `agency_q1/q2/q3` | Individual agency Likert items (1–10) |
| `agency_aggregate` | Mean of the three items |
| `ueqs_pragmatic` | UEQ-S pragmatic quality subscale |
| `ari_immersion` | ARI immersion score |
| `influence_score` | Continuous gait-change score from weight analysis |
| `category` | Classified response: *Consistent influence*, *Subconscious correction*, *No clear effect* |
| `direction` | Weight manipulation direction (*increasing* / *decreasing*) |

## Analysis plan
1. **Descriptive overview** — distributions and summary stats
2. **Correlation matrix** — Spearman ρ between all agency/experience metrics and `influence_score`
3. **Scatter plots** — agency aggregate vs influence score, broken down by direction
4. **Group comparison** — agency scores by influence category (boxplots + Kruskal-Wallis)
5. **Individual agency items** — check whether any single item drives the pattern

```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy import stats
from itertools import combinations

%matplotlib inline
plt.rcParams.update({"figure.dpi": 120, "font.size": 10})

DATA_DIR  = Path(".")
EXPORT_DIR = DATA_DIR / "exports"

df = pd.read_csv(EXPORT_DIR / "weight_influence_table.csv")
df["pid"] = df["pid"].astype(str).str.zfill(2)

AGENCY_COLS = ["agency_q1", "agency_q2", "agency_q3", "agency_aggregate",
               "ueqs_pragmatic", "ari_immersion"]
CATEGORY_ORDER = ["No clear effect", "Subconscious correction", "Consistent influence"]
CATEGORY_COLORS = {
    "No clear effect":        "#7f8c8d",
    "Subconscious correction": "#e67e22",
    "Consistent influence":    "#27ae60",
}
DIRECTION_COLORS = {"increasing": "#2980b9", "decreasing": "#c0392b"}

print(f"Loaded {len(df)} participants.")
print(f"Influence categories: {df['category'].value_counts().to_dict()}")
print(f"Directions: {df['direction'].value_counts().to_dict()}")
```

## 1 · Descriptive overview

```python
desc = df[AGENCY_COLS + ["influence_score"]].describe().T
desc.index.name = "variable"
display(desc.round(3))
```

```python
fig, axes = plt.subplots(2, 4, figsize=(14, 6))
fig.suptitle("Distributions of agency / experience metrics and influence score", fontsize=11, y=1.01)

plot_vars = AGENCY_COLS + ["influence_score"]
var_labels = {
    "agency_q1": "Agency Q1",
    "agency_q2": "Agency Q2",
    "agency_q3": "Agency Q3",
    "agency_aggregate": "Agency (aggregate)",
    "ueqs_pragmatic": "UEQ-S Pragmatic",
    "ari_immersion": "ARI Immersion",
    "influence_score": "Influence Score",
}

for ax, var in zip(axes.flat, plot_vars):
    vals = df[var].dropna()
    ax.hist(vals, bins=8, color="#3498db", edgecolor="white", alpha=0.85)
    ax.axvline(vals.median(), color="#e74c3c", lw=1.5, linestyle="--", label=f"median={vals.median():.2f}")
    ax.set_title(var_labels[var], fontsize=9)
    ax.legend(fontsize=7, frameon=False)

# hide the spare subplot
for ax in axes.flat[len(plot_vars):]:
    ax.set_visible(False)

plt.tight_layout()
plt.show()
```

## 2 · Spearman correlation matrix

```python
corr_vars = AGENCY_COLS + ["influence_score"]
rho_mat  = pd.DataFrame(index=corr_vars, columns=corr_vars, dtype=float)
pval_mat = pd.DataFrame(index=corr_vars, columns=corr_vars, dtype=float)

for v1 in corr_vars:
    for v2 in corr_vars:
        valid = df[[v1, v2]].dropna()
        if len(valid) < 5 or v1 == v2:
            rho_mat.loc[v1, v2]  = 1.0 if v1 == v2 else np.nan
            pval_mat.loc[v1, v2] = np.nan
            continue
        rho, p = stats.spearmanr(valid[v1], valid[v2])
        rho_mat.loc[v1, v2]  = rho
        pval_mat.loc[v1, v2] = p

# ── Heatmap ──────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 6))
rho_arr = rho_mat.values.astype(float)
im = ax.imshow(rho_arr, vmin=-1, vmax=1, cmap="RdBu_r", aspect="auto")
plt.colorbar(im, ax=ax, label="Spearman ρ")

tick_labels = [
    "Agency Q1", "Agency Q2", "Agency Q3",
    "Agency (agg)", "UEQ-S Prag.", "ARI Imm.", "Influence"
]
ax.set_xticks(range(len(corr_vars))); ax.set_xticklabels(tick_labels, rotation=35, ha="right", fontsize=9)
ax.set_yticks(range(len(corr_vars))); ax.set_yticklabels(tick_labels, fontsize=9)

for i in range(len(corr_vars)):
    for j in range(len(corr_vars)):
        r   = rho_arr[i, j]
        p   = pval_mat.values[i, j]
        if np.isnan(r): continue
        sig = "**" if (not np.isnan(p) and p < 0.01) else ("*" if (not np.isnan(p) and p < 0.05) else "")
        txt = f"{r:.2f}{sig}"
        col = "white" if abs(r) > 0.5 else "black"
        ax.text(j, i, txt, ha="center", va="center", fontsize=7, color=col)

ax.set_title("Spearman correlation matrix\n(* p<.05, ** p<.01)", fontsize=10)
plt.tight_layout()
plt.show()

# ── Focused row: correlations with influence_score ───────────────────────────
print("\nCorrelations with influence_score:")
for v in AGENCY_COLS:
    valid = df[[v, "influence_score"]].dropna()
    rho, p = stats.spearmanr(valid[v], valid["influence_score"])
    sig = "**" if p < 0.01 else ("*" if p < 0.05 else "ns")
    print(f"  {v:<22s}  ρ = {rho:+.3f}  p = {p:.3f}  {sig}  (n={len(valid)})")
```

## 3 · Scatter plots — agency aggregate vs influence score

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4.5), sharey=True)
fig.suptitle("Agency aggregate vs influence score", fontsize=11)

subsets = [
    ("All participants",         df),
    ("Increasing direction",     df[df["direction"] == "increasing"]),
    ("Decreasing direction",     df[df["direction"] == "decreasing"]),
]

for ax, (title, sub) in zip(axes, subsets):
    sub = sub.dropna(subset=["agency_aggregate", "influence_score"])
    if len(sub) < 3:
        ax.set_title(title + " (too few)"); continue

    # colour by category
    for cat, grp in sub.groupby("category"):
        ax.scatter(grp["agency_aggregate"], grp["influence_score"],
                   color=CATEGORY_COLORS.get(cat, "#95a5a6"),
                   label=cat, s=60, zorder=3, edgecolors="white", linewidths=0.5)

    # label each point with pid
    for _, row in sub.iterrows():
        ax.annotate(row["pid"], (row["agency_aggregate"], row["influence_score"]),
                    fontsize=6, ha="left", va="bottom",
                    xytext=(3, 2), textcoords="offset points", color="#555")

    # regression line
    x, y = sub["agency_aggregate"].values, sub["influence_score"].values
    m, b, r, p_lin, _ = stats.linregress(x, y)
    xfit = np.linspace(x.min(), x.max(), 100)
    ax.plot(xfit, m * xfit + b, color="#2c3e50", lw=1.2, linestyle="--",
            label=f"r={r:.2f}, p={p_lin:.3f}")

    rho, p_sp = stats.spearmanr(x, y)
    ax.set_title(f"{title}\nSpearman ρ={rho:.2f}, p={p_sp:.3f}", fontsize=9)
    ax.set_xlabel("Agency aggregate")
    ax.legend(fontsize=7, frameon=True, loc="upper right")

axes[0].set_ylabel("Influence score")
plt.tight_layout()
plt.show()
```

## 4 · Group comparison — agency by influence category

```python
agency_metrics = {
    "agency_aggregate": "Agency (aggregate)",
    "ueqs_pragmatic":   "UEQ-S Pragmatic",
    "ari_immersion":    "ARI Immersion",
}

fig, axes = plt.subplots(1, 3, figsize=(13, 5))
fig.suptitle("Agency / experience scores by influence category", fontsize=11)

for ax, (col, label) in zip(axes, agency_metrics.items()):
    groups = [df[df["category"] == cat][col].dropna().values for cat in CATEGORY_ORDER]
    positions = range(len(CATEGORY_ORDER))

    bp = ax.boxplot(groups, positions=list(positions), patch_artist=True,
                    widths=0.5, medianprops=dict(color="black", lw=2))

    for patch, cat in zip(bp["boxes"], CATEGORY_ORDER):
        patch.set_facecolor(CATEGORY_COLORS[cat])
        patch.set_alpha(0.7)

    # overlay jittered raw points
    for i, (grp, cat) in enumerate(zip(groups, CATEGORY_ORDER)):
        jitter = np.random.default_rng(i).uniform(-0.12, 0.12, size=len(grp))
        ax.scatter(i + jitter, grp, color=CATEGORY_COLORS[cat],
                   s=30, zorder=5, edgecolors="white", linewidths=0.5, alpha=0.9)

    # Kruskal-Wallis
    kw_stat, kw_p = stats.kruskal(*[g for g in groups if len(g) > 0])
    ax.set_title(f"{label}\nKruskal-Wallis H={kw_stat:.2f}, p={kw_p:.3f}", fontsize=9)
    ax.set_xticks(list(positions))
    short_labels = ["No effect", "Subconscious", "Consistent"]
    ax.set_xticklabels(short_labels, fontsize=8)
    ax.set_ylabel(label)

    # pairwise Mann-Whitney with annotation
    y_max = max(max(g) for g in groups if len(g) > 0)
    y_step = (y_max - ax.get_ylim()[0]) * 0.08
    for (i, j) in combinations(range(len(CATEGORY_ORDER)), 2):
        if len(groups[i]) < 2 or len(groups[j]) < 2: continue
        _, p_mw = stats.mannwhitneyu(groups[i], groups[j], alternative="two-sided")
        if p_mw < 0.10:
            sig = "**" if p_mw < 0.01 else ("*" if p_mw < 0.05 else "†")
            y_ann = y_max + y_step * (1 + abs(i - j))
            ax.plot([i, j], [y_ann, y_ann], color="#2c3e50", lw=0.8)
            ax.text((i + j) / 2, y_ann + y_step * 0.3, sig,
                    ha="center", fontsize=8, color="#2c3e50")

plt.tight_layout()
plt.show()

# ── Summary table ─────────────────────────────────────────────────────────────
print("\nMean ± SD by category:")
summary = df.groupby("category")[list(agency_metrics.keys())].agg(["mean", "std", "count"])
display(summary.round(2))
```

## 5 · Individual agency items

```python
q_labels = {"agency_q1": "Q1 (control)", "agency_q2": "Q2 (intention)", "agency_q3": "Q3 (voluntary)"}

fig, axes = plt.subplots(1, 3, figsize=(13, 4.5))
fig.suptitle("Individual agency items vs influence score", fontsize=11)

for ax, (col, label) in zip(axes, q_labels.items()):
    sub = df.dropna(subset=[col, "influence_score"])

    for cat, grp in sub.groupby("category"):
        ax.scatter(grp[col], grp["influence_score"],
                   color=CATEGORY_COLORS.get(cat, "#95a5a6"),
                   label=cat, s=55, zorder=3, edgecolors="white", linewidths=0.5)

    for _, row in sub.iterrows():
        ax.annotate(row["pid"], (row[col], row["influence_score"]),
                    fontsize=6, ha="left", va="bottom",
                    xytext=(3, 2), textcoords="offset points", color="#555")

    x, y = sub[col].values.astype(float), sub["influence_score"].values
    rho, p_sp = stats.spearmanr(x, y)
    m, b, r, p_lin, _ = stats.linregress(x, y)
    xfit = np.linspace(x.min(), x.max(), 100)
    ax.plot(xfit, m * xfit + b, color="#2c3e50", lw=1.2, linestyle="--")

    ax.set_title(f"{label}\nSpearman ρ={rho:.2f}, p={p_sp:.3f}", fontsize=9)
    ax.set_xlabel(label)
    ax.set_ylabel("Influence score")
    ax.legend(fontsize=7, frameon=True)

plt.tight_layout()
plt.show()

# ── Print correlations table ──────────────────────────────────────────────────
print("\nSpearman correlations with influence_score (all agency items + experience):")
print(f"{'Variable':<24} {'ρ':>6}  {'p':>7}  {'sig':>4}  {'n':>3}")
print("-" * 50)
for col in list(q_labels.keys()) + list(agency_metrics.keys()):
    valid = df[[col, "influence_score"]].dropna()
    rho, p = stats.spearmanr(valid[col], valid["influence_score"])
    sig = "**" if p < 0.01 else ("*" if p < 0.05 else "†" if p < 0.10 else "ns")
    lbl = q_labels.get(col, agency_metrics.get(col, col))
    print(f"{lbl:<24} {rho:+.3f}  {p:.4f}  {sig:>4}  {len(valid):>3}")
```

## 6 · Direction × agency interaction

Check whether the agency–influence relationship differs between the *increasing* and *decreasing* weight conditions.

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 5), sharey=True)
fig.suptitle("Agency aggregate vs influence score — split by weight direction", fontsize=11)

for ax, direction in zip(axes, ["increasing", "decreasing"]):
    sub = df[df["direction"] == direction].dropna(subset=["agency_aggregate", "influence_score"])

    for cat, grp in sub.groupby("category"):
        ax.scatter(grp["agency_aggregate"], grp["influence_score"],
                   color=CATEGORY_COLORS.get(cat, "#95a5a6"),
                   label=cat, s=70, zorder=3, edgecolors="white", linewidths=0.5)

    for _, row in sub.iterrows():
        ax.annotate(row["pid"], (row["agency_aggregate"], row["influence_score"]),
                    fontsize=7, ha="left", va="bottom",
                    xytext=(3, 2), textcoords="offset points", color="#555")

    if len(sub) >= 3:
        x, y = sub["agency_aggregate"].values, sub["influence_score"].values
        m, b, r, p_lin, _ = stats.linregress(x, y)
        xfit = np.linspace(x.min(), x.max(), 100)
        ax.plot(xfit, m * xfit + b, color=DIRECTION_COLORS[direction],
                lw=1.5, linestyle="--", label=f"r={r:.2f}")
        rho, p_sp = stats.spearmanr(x, y)
        ax.set_title(f"{direction.capitalize()} (n={len(sub)})\nSpearman ρ={rho:.2f}, p={p_sp:.3f}", fontsize=9)

    ax.set_xlabel("Agency aggregate")
    ax.legend(fontsize=7, frameon=True)

axes[0].set_ylabel("Influence score")
plt.tight_layout()
plt.show()

# ── Compare mean agency between directions ─────────────────────────────────────
print("\nAgency aggregate by direction:")
for d in ["increasing", "decreasing"]:
    vals = df[df["direction"] == d]["agency_aggregate"].dropna()
    print(f"  {d:<12}  mean={vals.mean():.2f}  sd={vals.std():.2f}  n={len(vals)}")

inc = df[df["direction"] == "increasing"]["agency_aggregate"].dropna()
dec = df[df["direction"] == "decreasing"]["agency_aggregate"].dropna()
u, p_dir = stats.mannwhitneyu(inc, dec, alternative="two-sided")
print(f"\nMann-Whitney U={u:.0f}, p={p_dir:.3f} (direction difference in agency)")
```

## Summary

Run the cells above to populate the findings. Key things to look for:

- **Positive ρ** between agency and influence → high-agency participants show *more* gait adaptation to the weight sound
- **Negative ρ** → high-agency participants resist the manipulation more
- **Category differences** → whether the *Consistent influence* group scores systematically higher/lower on any agency item
- **Direction interaction** → whether increasing vs decreasing weight produces a different agency–influence pattern
